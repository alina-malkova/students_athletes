```python
import numpy as np

# 1. Shows all column names that contain 'NRAL' in 'ef'
nral_cols = [col for col in ef.columns if 'NRAL' in col.upper()]
print("Columns containing 'NRAL' in 'ef' DataFrame:")
print(nral_cols)

# Filter out 'X' columns which usually indicate imputation flags
nral_enrollment_cols = [col for col in nral_cols if not col.startswith('X')]
print("\nNon-imputation 'NRAL' enrollment columns:")
print(nral_enrollment_cols)

# Convert relevant columns to numeric, coercing errors
for col in nral_enrollment_cols:
    ef[col] = pd.to_numeric(ef[col], errors='coerce')

# 2. Calculates the total nonresident alien enrollment across all institutions
# Summing across relevant columns for each institution (UNITID)
# Fill NaN with 0 before summing to ensure correct total, as NaN means no data/missing
ef['TOTAL_NRAL_ENROLLMENT'] = ef[nral_enrollment_cols].sum(axis=1)

# Aggregate total enrollment by UNITID
total_nral_by_unitid = ef.groupby('UNITID')['TOTAL_NRAL_ENROLLMENT'].sum().reset_index()

print(f"\nTotal Nonresident Alien Enrollment calculated for {len(total_nral_by_unitid)} institutions.")
```

```python
import os
import pandas as pd

# Define the directory path (already available from previous cells)
base_path = '/content/drive/MyDrive/RA_Project_EconDistance/raw_data/'

# 1. List all files in the base_path directory
all_files = os.listdir(base_path)
print(f"All files in '{base_path}':")
for f in all_files:
    print(f"- {f}")

# 2. Filter this list to include only files that start with 'WDI_' and end with '.csv'
wdi_csv_files = [f for f in all_files if f.startswith('WDI_') and f.endswith('.csv')]
print(f"\nFiltered WDI CSV files: {wdi_csv_files}")

# 3. Initialize an empty dictionary to store the loaded DataFrames
wdi_dataframes = {}

# 4. Loop through each filtered WDI CSV file
print("\n--- Loading WDI DataFrames ---")
for filename in wdi_csv_files:
    file_path = os.path.join(base_path, filename)

    try:
        # b. Load the CSV file into a pandas DataFrame
        df = pd.read_csv(file_path)

        # c. Extract a descriptive name for the DataFrame from the filename
        # Remove 'WDI_' prefix and '.csv' suffix
        descriptive_name = filename.replace('WDI_', '').replace('.csv', '')

        # d. Store the loaded DataFrame in the wdi_dataframes dictionary
        wdi_dataframes[descriptive_name] = df

        # e. Print verification information
        print(f"\nLoaded File: {filename}")
        print(f"DataFrame Name: {descriptive_name}")
        print(f"Shape: {df.shape}")
        print(f"First 5 Columns: {list(df.columns[:5])}")
        print("-" * 50)

    except Exception as e:
        print(f"Error loading {filename}: {e}")

print("\nAll WDI CSV files processed and loaded into 'wdi_dataframes' dictionary.")
print(f"Number of WDI DataFrames loaded: {len(wdi_dataframes)}")
print(f"Keys in wdi_dataframes: {list(wdi_dataframes.keys())}")
```

```python
import pandas as pd

# 1. Create a mapping dictionary to convert the Country names to their corresponding economy (country) codes.
# Using df_indicator from kernel state for country code mapping
# The index of df_indicator is the economy code, and 'Country' column has the name.
# Ensure to reset index to access 'economy' as a column, then create the mapping.
country_name_to_code = df_indicator.reset_index().set_index('Country')['economy'].to_dict()

print(f"Created country_name_to_code mapping for {len(country_name_to_code)} countries.")
print("Sample of mapping:", dict(list(country_name_to_code.items())[:5]))

# 2. Initialize an empty list called reshaped_wdi_dfs
reshaped_wdi_dfs = []

# 3. Define a dictionary, indicator_rename_map, to map descriptive indicator names to shorter column names
indicator_rename_map = {
    'PPP_conversion_factor': 'ppp',
    'Official_exchange_rate': 'exchange_rate',
    'Unemployment_rate': 'unemployment',
    'GDP_per_capita_growth': 'gdp_growth',
    'GDP_per_capita_PPP': 'gdp_per_capita_ppp',
    'Price_level_ratio': 'price_level_ratio'
}

# 4. Iterate through each item (descriptive indicator name and DataFrame) in the wdi_dataframes dictionary:
print("\n--- Reshaping WDI DataFrames ---")
for descriptive_indicator_name, df_wdi in wdi_dataframes.items():
    # a. Create a copy of the current DataFrame to avoid modifying the original.
    df_copy = df_wdi.copy()

    # b. Add a new column named `country_code` to the copied DataFrame by mapping the values
    #    in its 'Country' column using the `country_name_to_code` dictionary.
    #    Use fillna('Unknown') for any country names not found in the mapping.
    df_copy['country_code'] = df_copy['Country'].map(country_name_to_code).fillna('Unknown')

    # c. Rename the 'Country' column to 'country_name'.
    df_copy.rename(columns={'Country': 'country_name'}, inplace=True)

    # d. Identify all columns that start with 'YR' as the year columns.
    year_cols = [col for col in df_copy.columns if col.startswith('YR')]

    # e. Use the pd.melt() function to transform the DataFrame from wide to long format.
    #    Set id_vars=['country_code', 'country_name'], value_vars to the identified year columns,
    #    var_name='year', and value_name to the corresponding new indicator name from indicator_rename_map.
    new_indicator_name = indicator_rename_map.get(descriptive_indicator_name, descriptive_indicator_name) # Fallback to original name if not mapped
    df_long = pd.melt(df_copy,
                      id_vars=['country_code', 'country_name'],
                      value_vars=year_cols,
                      var_name='year',
                      value_name=new_indicator_name)

    # f. Convert the 'year' column from string format (e.g., 'YR2010') to integer format (e.g., 2010)
    #    by removing the 'YR' prefix and casting to `int`.
    df_long['year'] = df_long['year'].str.replace('YR', '').astype(int)

    # g. Append the reshaped and cleaned DataFrame to the `reshaped_wdi_dfs` list.
    reshaped_wdi_dfs.append(df_long)
    print(f"Reshaped '{descriptive_indicator_name}' to long format with shape {df_long.shape}.")

# 5. After the loop, print the number of reshaped DataFrames and the first 5 rows of the first DataFrame
#    in reshaped_wdi_dfs to verify the transformation.
print(f"\nNumber of reshaped DataFrames: {len(reshaped_wdi_dfs)}")
if reshaped_wdi_dfs:
    print("First 5 rows of the first reshaped DataFrame:")
    display(reshaped_wdi_dfs[0].head())
```

```python
print("Cleaning institution name columns and performing fuzzy matching to create crosswalk...")

def clean_text(text):
    if pd.isna(text):
        return ""
    text = str(text).lower() # Convert to lowercase
    text = text.translate(str.maketrans('', '', string.punctuation)) # Remove punctuation
    text = ' '.join(text.split()) # Remove extra spaces by splitting and joining
    return text

# Ensure 'cleaned_name' column exists in ncaa DataFrame
# If it already exists, this will re-clean it, which is safe.
if 'cleaned_name' not in ncaa.columns:
    ncaa['cleaned_name'] = ncaa['name'].apply(clean_text)
else:
    print("'cleaned_name' column already exists in 'ncaa'. Re-cleaning to ensure consistency.")
    ncaa['cleaned_name'] = ncaa['name'].apply(clean_text)

# Ensure 'INSTNM_cleaned' column exists in hd DataFrame
# If it already exists, this will re-clean it, which is safe.
if 'INSTNM_cleaned' not in hd.columns:
    hd['INSTNM_cleaned'] = hd['INSTNM'].apply(clean_text)
else:
    print("'INSTNM_cleaned' column already exists in 'hd'. Re-cleaning to ensure consistency.")
    hd['INSTNM_cleaned'] = hd['INSTNM'].apply(clean_text)

# Initialize an empty list to store the crosswalk results
crosswalk_results = []

# Define a threshold for low match scores to flag for review
LOW_SCORE_THRESHOLD = 75
low_score_matches = []

# Iterate through each institution in the NCAA DataFrame
for index_ncaa, row_ncaa in ncaa.iterrows():
    ncaa_name = row_ncaa['cleaned_name']
    best_match_score = 0
    best_match_hd_name = None
    best_match_unitid = None

    # Iterate through each institution in the HD (IPEDS) DataFrame to find the best match
    for index_hd, row_hd in hd.iterrows():
        hd_name = row_hd['INSTNM_cleaned']
        # Use token_set_ratio for better matching with differing word orders/extra words
        score = fuzz.token_set_ratio(ncaa_name, hd_name)

        if score > best_match_score:
            best_match_score = score
            best_match_hd_name = row_hd['INSTNM']
            best_match_unitid = row_hd['UNITID']

    # Store the result, including 'cleaned_name'
    crosswalk_results.append({
        'ncaa_name': row_ncaa['name'],
        'ncaa_state': row_ncaa['state'],
        'ncaa_division': row_ncaa['division'],
        'ncaa_conference': row_ncaa['conference'],
        'ipeds_name_matched': best_match_hd_name,
        'ipeds_unitid': best_match_unitid,
        'match_score': best_match_score,
        'cleaned_name': row_ncaa['cleaned_name'] # Add cleaned_name here
    })

    # Flag low-scoring matches for manual review
    if best_match_score < LOW_SCORE_THRESHOLD:
        low_score_matches.append({
            'ncaa_name': row_ncaa['name'],
            'ipeds_name_matched': best_match_hd_name,
            'match_score': best_match_score
        })

# Create the crosswalk DataFrame
crosswalk_df = pd.DataFrame(crosswalk_results)

print("Fuzzy matching complete.")

print(f"\nCrosswalk created with {len(crosswalk_df)} entries.")
print("First 5 rows of the crosswalk DataFrame:")
display(crosswalk_df.head())

print(f"\nNumber of low-scoring matches (< {LOW_SCORE_THRESHOLD}): {len(low_score_matches)}")
if low_score_matches:
    print("Top 10 low-scoring matches for review:")
    low_score_df = pd.DataFrame(low_score_matches).sort_values(by='match_score', ascending=True)
    display(low_score_df.head(10))

# Save the crosswalk to CSV
save_path = "/content/drive/MyDrive/RA_Project_EconDistance/raw_data/ncaa_ipeds_crosswalk.csv"
os.makedirs(os.path.dirname(save_path), exist_ok=True)
crosswalk_df.to_csv(save_path, index=False)
print(f"\nCrosswalk saved to: {save_path}")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
import os

# ============================================================
# CONFIGURATION
# ============================================================
HEADERS = {
    "User-Agent":      "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 "
                       "(KHTML, like Gecko) Chrome/124.0.0.0 Safari/537.36",
    "Accept":          "text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8",
    "Accept-Language": "en-US,en;q=0.9",
    "Referer":         "https://www.tfrrs.org/",
}

DELAY         = 3  # seconds between requests
RAW_DATA_PATH = "/content/drive/MyDrive/RA_Project_EconDistance/raw_data"

YEAR_MAP = {
    'SR-4': 'Senior',
    'JR-3': 'Junior',
    'SO-2': 'Sophomore',
    'FR-1': 'Freshman'
}

# ============================================================
# 1. LOAD TEAM URL LIST
# ============================================================
teams_df = pd.read_csv(f"{RAW_DATA_PATH}/tfrrs_team_urls.csv")
print(f"Loaded {len(teams_df)} team URLs")
display(teams_df.head(3))

# ============================================================
# 2. PARSING FUNCTION (same logic as single-page test)
# ============================================================
def scrape_roster(url, school_name, gender):
    """
    Scrapes the ROSTER table from a TFRRS team page.
    Returns a list of dicts, one per athlete.
    Raises an exception if the page or table cannot be parsed.
    """
    response = requests.get(url, headers=HEADERS, timeout=30)
    response.raise_for_status()

    soup = BeautifulSoup(response.text, 'html.parser')

    # Find ROSTER section
    roster_table = None
    for h3 in soup.find_all('h3'):
        if 'ROSTER' in h3.text.upper():
            roster_table = h3.find_next('table')
            break

    if roster_table is None:
        raise ValueError("ROSTER table not found on page")

    athletes = []
    for row in roster_table.find_all('tr')[1:]:  # skip header
        cols = row.find_all('td')
        if len(cols) < 2:
            continue

        name_tag = cols[0].find('a')
        name     = name_tag.text.strip() if name_tag else cols[0].text.strip()
        if not name:
            continue

        profile_url = ("https://www.tfrrs.org" + name_tag['href']
                       if name_tag and name_tag.get('href') else None)

        year_raw   = cols[1].text.strip()
        class_year = YEAR_MAP.get(year_raw, year_raw)

        athletes.append({
            'athlete_name': name,
            'class_year':   class_year,
            'hometown':     None,   # only on individual athlete pages
            'event':        None,   # only on individual athlete pages
            'school_name':  school_name,
            'gender':       gender,
            'sport':        'Track & Field',
            'profile_url':  profile_url,
            'team_url':     url,
        })

    return athletes

# ============================================================
# 3. MAIN LOOP
# ============================================================
all_athletes = []
failed_teams = []
success_count = 0
total = len(teams_df)

print(f"\nStarting scrape of {total} teams...\n")

for i, row in teams_df.iterrows():
    url         = row['tfrrs_team_url']
    school_name = row['school_name']
    gender      = row.get('gender', 'Unknown')

    try:
        athletes = scrape_roster(url, school_name, gender)

        if len(athletes) == 0:
            raise ValueError("Roster parsed but 0 athletes found")

        all_athletes.extend(athletes)
        success_count += 1

    except Exception as e:
        failed_teams.append({
            'school_name': school_name,
            'gender':      gender,
            'team_url':    url,
            'error':       str(e),
        })

    # Progress update every 10 teams
    teams_done = i + 1
    if teams_done % 10 == 0 or teams_done == total:
        print(f"  Scraped {teams_done}/{total} | "
              f"✅ {success_count} success | "
              f"❌ {len(failed_teams)} failed | "
              f"Athletes so far: {len(all_athletes)}")

    time.sleep(DELAY)

# ============================================================
# 4. COMBINE INTO DATAFRAME
# ============================================================
roster_df = pd.DataFrame(all_athletes)
failures_df = pd.DataFrame(failed_teams)

# ============================================================
# 5. SUMMARY
# ============================================================
print(f"\n{'='*55}")
print(f"SCRAPING COMPLETE")
print(f"{'='*55}")
print(f"Total athletes scraped:      {len(roster_df)}")
print(f"Teams successfully scraped:  {success_count}")
print(f"Teams failed:                {len(failed_teams)}")
if len(failed_teams) > 0:
    print(f"\nFailed teams:")
    display(failures_df[['school_name', 'gender', 'error']])

# ============================================================
# 6. SAVE OUTPUTS
# ============================================================
os.makedirs(RAW_DATA_PATH, exist_ok=True)

roster_path   = f"{RAW_DATA_PATH}/tfrrs_all_rosters.csv"
failures_path = f"{RAW_DATA_PATH}/tfrrs_failures.csv"

roster_df.to_csv(roster_path, index=False)
print(f"\n✅ Full roster saved to:  {roster_path}")

failures_df.to_csv(failures_path, index=False)
print(f"✅ Failures log saved to: {failures_path}")

print(f"\nFirst 20 rows of roster:")
display(roster_df.head(20))
```

```python
import subprocess
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
import os

RAW_DATA_PATH = "/content/drive/MyDrive/RA_Project_EconDistance/raw_data"

HEADERS = {
    "User-Agent":      "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 "
                       "(KHTML, like Gecko) Chrome/124.0.0.0 Safari/537.36",
    "Accept":          "text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8",
    "Accept-Language": "en-US,en;q=0.9",
    "Referer":         "https://www.tfrrs.org/",
}

YEAR_MAP = {
    'SR-4': 'Senior',
    'JR-3': 'Junior',
    'SO-2': 'Sophomore',
    'FR-1': 'Freshman'
}

# ============================================================
# 1. LOAD FAILURES
# ============================================================
failures_df = pd.read_csv(f"{RAW_DATA_PATH}/tfrrs_failures.csv")
print(f"Failed URLs to retry: {len(failures_df)}")
display(failures_df)

# ============================================================
# 2. PLAYWRIGHT FETCH FUNCTION
#    Uses a real browser to render JS — saves HTML to /tmp
# ============================================================
def fetch_with_playwright(url):
    """Fetches a URL using Playwright headless browser, returns HTML string."""
    script = f"""
import asyncio
from playwright.async_api import async_playwright

async def main():
    async with async_playwright() as p:
        browser = await p.chromium.launch(
            headless=True,
            args=["--no-sandbox", "--disable-dev-shm-usage"]
        )
        page = await browser.new_page(
            user_agent="Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/124.0.0.0 Safari/537.36"
        )
        await page.goto("{url}", wait_until="domcontentloaded", timeout=60000)
        await page.wait_for_timeout(5000)
        html = await page.content()
        await browser.close()
        with open("/tmp/retry_page.html", "w") as f:
            f.write(html)

asyncio.run(main())
"""
    with open("/tmp/retry_scrape.py", "w") as f:
        f.write(script)

    result = subprocess.run(
        ["python", "/tmp/retry_scrape.py"],
        capture_output=True, text=True, timeout=90
    )

    if result.returncode != 0:
        raise RuntimeError(f"Playwright failed: {result.stderr[:200]}")

    with open("/tmp/retry_page.html", "r") as f:
        return f.read()

# ============================================================
# 3. PARSE ROSTER FROM HTML
# ============================================================
def parse_roster(html, url, school_name, gender):
    soup = BeautifulSoup(html, 'html.parser')

    roster_table = None
    for h3 in soup.find_all('h3'):
        if 'ROSTER' in h3.text.upper():
            roster_table = h3.find_next('table')
            break

    if roster_table is None:
        raise ValueError("ROSTER table not found")

    athletes = []
    for row in roster_table.find_all('tr')[1:]:
        cols = row.find_all('td')
        if len(cols) < 2:
            continue

        name_tag = cols[0].find('a')
        name     = name_tag.text.strip() if name_tag else cols[0].text.strip()
        if not name:
            continue

        profile_url = ("https://www.tfrrs.org" + name_tag['href']
                       if name_tag and name_tag.get('href') else None)

        year_raw   = cols[1].text.strip()
        class_year = YEAR_MAP.get(year_raw, year_raw)

        athletes.append({
            'athlete_name': name,
            'class_year':   class_year,
            'hometown':     None,
            'event':        None,
            'school_name':  school_name,
            'gender':       gender,
            'sport':        'Track & Field',
            'profile_url':  profile_url,
            'team_url':     url,
        })

    if len(athletes) == 0:
        raise ValueError("0 athletes found after parsing")

    return athletes

# ============================================================
# 4. RETRY LOOP — requests first, Playwright as fallback
# ============================================================
recovered_athletes = []
still_failed       = []

total = len(failures_df)

for i, row in failures_df.iterrows():
    url         = row['team_url']
    school_name = row['school_name']
    gender      = row.get('gender', 'Unknown')
    attempt_num = i + 1

    print(f"\n[{attempt_num}/{total}] Retrying: {school_name} ({gender})")

    html = None

    # --- Try 1: plain requests (faster) ---
    try:
        response = requests.get(url, headers=HEADERS, timeout=30)
        response.raise_for_status()
        html = response.text
        print(f"  ✅ requests succeeded")
    except Exception as e:
        print(f"  ⚠️  requests failed ({e}) → trying Playwright...")

    # --- Try 2: Playwright if requests failed ---
    if html is None:
        try:
            html = fetch_with_playwright(url)
            print(f"  ✅ Playwright succeeded")
        except Exception as e:
            print(f"  ❌ Playwright also failed: {e}")
            still_failed.append({
                'school_name': school_name,
                'gender':      gender,
                'team_url':    url,
                'error':       str(e),
            })
            time.sleep(5)
            continue

    # --- Parse roster ---
    try:
        athletes = parse_roster(html, url, school_name, gender)
        recovered_athletes.extend(athletes)
        print(f"  ✅ Parsed {len(athletes)} athletes")
    except Exception as e:
        print(f"  ❌ Parse failed: {e}")
        still_failed.append({
            'school_name': school_name,
            'gender':      gender,
            'team_url':    url,
            'error':       f"Parse error: {e}",
        })

    time.sleep(5)

# ============================================================
# 5. APPEND RECOVERED DATA TO MAIN ROSTER
# ============================================================
print(f"\n{'='*55}")
print(f"RETRY COMPLETE")
print(f"  Recovered:    {len(failures_df) - len(still_failed)} teams "
      f"({len(recovered_athletes)} athletes)")
print(f"  Still failed: {len(still_failed)} teams")

if recovered_athletes:
    main_path  = f"{RAW_DATA_PATH}/tfrrs_all_rosters.csv"
    existing   = pd.read_csv(main_path) if os.path.exists(main_path) else pd.DataFrame()
    recovered  = pd.DataFrame(recovered_athletes)
    combined   = pd.concat([existing, recovered], ignore_index=True)
    combined.to_csv(main_path, index=False)
    print(f"\n✅ Main roster updated: {len(combined)} total athletes → {main_path}")

# ============================================================
# 6. SAVE STILL-FAILED URLs
# ============================================================
still_failed_df = pd.DataFrame(still_failed)
still_failed_path = f"{RAW_DATA_PATH}/tfrrs_still_failed.csv"
still_failed_df.to_csv(still_failed_path, index=False)
print(f"✅ Still-failed URLs saved to: {still_failed_path}")

if len(still_failed) > 0:
    print("\nURLs that still failed after retry:")
    display(still_failed_df[['school_name', 'gender', 'error']])
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
import os

RAW_DATA_PATH = "/content/drive/MyDrive/RA_Project_EconDistance/raw_data"
os.makedirs(RAW_DATA_PATH, exist_ok=True)

HEADERS = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 "
                  "(KHTML, like Gecko) Chrome/124.0.0.0 Safari/537.36",
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8",
}

# ============================================================
# SCHOOLS LIST
# Add more schools here as you expand
# Format: ("School Name", "https://athletics-url.com/")
# ============================================================
SCHOOLS = [
    ("University of Virginia",       "https://virginiasports.com/"),
    ("LSU",                          "https://lsusports.net/"),
    ("University of Michigan",       "https://mgoblue.com/"),
    ("University of Florida",        "https://floridagators.com/"),
    ("Stanford University",          "https://gostanford.com/"),
    ("MIT Athletics",                 "https://mitathletics.com/"),
    ("Denison Big Red",               "https://denisonbigred.com/"),
    ("University of North Carolina", "https://goheels.com/"),
    ("Duke University",              "https://goduke.com/"),
    ("Ohio State University",        "https://ohiostatebuckeyes.com/"),
    ("Georgetown University",        "https://guhoyas.com/"),
    ("University of Texas",          "https://texaslonghorns.com/"),
]

# ============================================================
# PLATFORM DETECTION — checks multiple signals per platform
# ============================================================
def detect_platform(url, html, soup):
    url_lower  = url.lower()
    html_lower = html.lower()

    # --- Collect all text signals ---
    # Meta tags (description, generator, author)
    meta_content = ' '.join([
        (tag.get('content') or '') for tag in soup.find_all('meta')
    ]).lower()

    # Script src attributes
    script_srcs = ' '.join([
        (tag.get('src') or '') for tag in soup.find_all('script')
    ]).lower()

    # Link href attributes
    link_hrefs = ' '.join([
        (tag.get('href') or '') for tag in soup.find_all('link')
    ]).lower()

    # All signals combined
    all_signals = ' '.join([
        url_lower, html_lower, meta_content, script_srcs, link_hrefs
    ])

    # --- SIDEARM detection ---
    sidearm_keywords = [
        'sidearm', 'sidearmsports', 'sidearm.nextgen',
        'sb.scorecardresearch',        # Sidearm always loads this
        'dxbhsrqyrr690.cloudfront',   # Sidearm CDN
        'sidearm-nav', 's-person-card',
    ]
    if any(kw in all_signals for kw in sidearm_keywords):
        return 'Sidearm Sports'

    # --- WMT detection ---
    wmt_keywords = [
        'wmtdigital', 'wmt digital', 'partner of wmt',
        'wmt-digital', 'wmtsports',
    ]
    if any(kw in all_signals for kw in wmt_keywords):
        return 'WMT Digital'

    # --- PRESTO detection ---
    presto_keywords = [
        'prestosports', 'presto sports', 'presto-sports',
        'prestoedge', 'presto.com',
    ]
    if any(kw in all_signals for kw in presto_keywords):
        return 'PrestoSports'

    return 'Unknown'

# ============================================================
# MAIN LOOP
# ============================================================
results = []
print(f"Detecting platforms for {len(SCHOOLS)} schools...\n")

for school_name, url in SCHOOLS:
    try:
        # Some schools redirect — follow redirects and use final URL
        response  = requests.get(url, headers=HEADERS,
                                 timeout=20, allow_redirects=True)
        final_url = response.url  # URL after redirects
        response.raise_for_status()

        soup     = BeautifulSoup(response.text, 'html.parser')
        platform = detect_platform(final_url, response.text, soup)
        status   = 'OK'

        # Flag if redirected to a different domain
        redirect_note = ''
        if final_url.rstrip('/') != url.rstrip('/'):
            redirect_note = f' (redirected → {final_url})'

        print(f"✅ {school_name:40} → {platform}{redirect_note}")

    except Exception as e:
        platform  = 'Error'
        final_url = url
        status    = str(e)[:100]
        print(f"❌ {school_name:40} → Error: {status}")

    results.append({
        'school_name': school_name,
        'url':         final_url,   # use final URL after redirect
        'platform':    platform,
        'status':      status,
    })
    time.sleep(2)

# ============================================================
# RESULTS
# ============================================================
platforms_df = pd.DataFrame(results)

print(f"\n{'='*50}")
print("PLATFORM DISTRIBUTION:")
print(platforms_df['platform'].value_counts().to_string())
print(f"\nFull results:")
display(platforms_df[['school_name', 'url', 'platform', 'status']])

# ============================================================
# SAVE
# ============================================================
save_path = f"{RAW_DATA_PATH}/school_platforms_base.csv"

# If file already exists, append new schools (avoid duplicates)
if os.path.exists(save_path):
    existing  = pd.read_csv(save_path)
    combined  = pd.concat([existing, platforms_df], ignore_index=True)
    combined  = combined.drop_duplicates(subset='school_name', keep='last')
    combined.to_csv(save_path, index=False)
    print(f"\n✅ Updated existing file — total schools: {len(combined)}")
else:
    platforms_df.to_csv(save_path, index=False)
    print(f"\n✅ Saved to: {save_path} ({len(platforms_df)} schools)")
```

```python
# Célula: Expandir CSV de plataformas com links de rosters por gênero (versão corrigida)
import pandas as pd
import requests
from bs4 import BeautifulSoup
import time
import os

# Caminhos
BASE_PATH = '/content/drive/MyDrive/RA_Project_EconDistance/raw_data/'
INPUT_FILE = os.path.join(BASE_PATH, 'school_platforms_base.csv')
OUTPUT_FILE = os.path.join(BASE_PATH, 'school_website_platforms.csv')

# Padrões de URL para cada gênero
patterns = {
    'Men': ['mens-soccer'],
    'Women': ['womens-soccer']
}

def is_valid_roster_page(url):
    """
    Verifica se a URL é uma página de roster válida.
    Retorna True se status 200 e o conteúdo contém elementos típicos de roster.
    """
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36'
    }
    try:
        response = requests.get(url, headers=headers, timeout=15, allow_redirects=True)
        if response.status_code != 200:
            return False

        soup = BeautifulSoup(response.content, 'html.parser')

        # Verificar se há elementos que indicam um roster
        # Tabelas com classe 'roster' ou 'sidearm-roster'
        if soup.find('table', class_=re.compile(r'roster', re.I)):
            return True
        # Lista de jogadores (Sidearm, Presto)
        if soup.find('ul', class_='sidearm-roster-players'):
            return True
        # Cards (WMT)
        if soup.find('div', class_=re.compile(r'roster-card|player-card', re.I)):
            return True
        # Qualquer div com classe contendo 'roster' e que tenha links de jogadores
        roster_containers = soup.find_all('div', class_=re.compile(r'roster', re.I))
        for container in roster_containers:
            if container.find('a', href=re.compile(r'/roster/')) or container.find('img', alt=re.compile(r'Headshot')):
                return True

        # Fallback: se o título ou o corpo contém a palavra "roster" (menos confiável)
        title = soup.title.string.lower() if soup.title else ''
        if 'roster' in title:
            return True

        # Se chegou aqui, provavelmente não é um roster
        return False

    except Exception as e:
        print(f"    ⚠️ Erro ao verificar URL: {e}")
        return False

# Carregar CSV
print(f"📂 Lendo {INPUT_FILE}...")
df = pd.read_csv(INPUT_FILE)

expanded_rows = []

for idx, row in df.iterrows():
    school = row['school_name']
    base_url = row['url'].strip()
    platform = row['platform']
    status = row.get('status', 'OK')

    if not base_url.endswith('/'):
        base_url += '/'

    print(f"\n[{idx+1}/{len(df)}] {school} ({platform})")

    for gender, pat_list in patterns.items():
        roster_url = None
        for pattern in pat_list:
            candidate = base_url + f'sports/{pattern}/roster'
            print(f"  Testando {gender}: {candidate}")
            if is_valid_roster_page(candidate):
                roster_url = candidate
                print(f"    ✅ Válido! (página de roster encontrada)")
                break
            else:
                print(f"    ❌ Inválido (página não contém roster)")
        # Adicionar linha
        expanded_rows.append({
            'school_name': school,
            'gender': gender,
            'platform': platform,
            'base_url': base_url,
            'roster_url': roster_url,
            'status': status
        })
        time.sleep(1)  # delay entre requisições

# Criar DataFrame expandido
df_expanded = pd.DataFrame(expanded_rows)

# Salvar
df_expanded.to_csv(OUTPUT_FILE, index=False, encoding='utf-8')
print(f"\n✅ Arquivo salvo em: {OUTPUT_FILE}")
print(f"Total de linhas: {len(df_expanded)}")
print(df_expanded.head())
```

```python
# Célula 4: Função para extrair dados do roster do Sidearm Sports (VERSÃO ALTERNATIVA SEM SELENIUM)
def scrape_sidearm_roster_requests(url, school_name):
    """
    Extrai dados do roster usando requests e BeautifulSoup (alternativa ao Selenium)

    Args:
        url (str): URL da página do roster
        school_name (str): Nome da faculdade/universidade

    Returns:
        pandas.DataFrame: DataFrame com os dados dos atletas
    """
    import requests
    import json
    import re

    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36',
        'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8',
        'Accept-Language': 'en-US,en;q=0.5',
        'Accept-Encoding': 'gzip, deflate, br',
        'Connection': 'keep-alive',
        'Upgrade-Insecure-Requests': '1'
    }

    roster_data = []

    try:
        print(f"🌐 Acessando: {url}")
        response = requests.get(url, headers=headers, timeout=30)
        response.raise_for_status()

        soup = BeautifulSoup(response.content, 'html.parser')

        # Procurar por dados JSON-LD ou script que contenha os dados
        scripts = soup.find_all('script', type='application/ld+json')
        for script in scripts:
            try:
                data = json.loads(script.string)
                if data.get('@type') == 'SportsTeam' or 'athlete' in str(data).lower():
                    print("✅ Encontrados dados estruturados JSON-LD")
                    # Processar dados estruturados
                    if 'athlete' in data:
                        for athlete in data['athlete']:
                            roster_data.append({
                                'school_name': school_name,
                                'url': url,
                                'jersey_number': athlete.get('identifier', ''),
                                'name': athlete.get('name', ''),
                                'position': athlete.get('position', ''),
                                'class_year': '',
                                'hometown': '',
                                'height': '',
                                'high_school_club': ''
                            })
            except:
                pass

        # Método 1: Tabela tradicional
        if not roster_data:
            # Procurar por tabelas
            tables = soup.find_all('table')
            for table in tables:
                if 'roster' in str(table).lower():
                    rows = table.find_all('tr')
                    for row in rows[1:]:  # Pular cabeçalho
                        cols = row.find_all('td')
                        if len(cols) >= 4:
                            athlete = {
                                'school_name': school_name,
                                'url': url,
                                'jersey_number': cols[0].get_text(strip=True) if len(cols) > 0 else '',
                                'name': cols[1].get_text(strip=True) if len(cols) > 1 else '',
                                'position': cols[2].get_text(strip=True) if len(cols) > 2 else '',
                                'class_year': cols[3].get_text(strip=True) if len(cols) > 3 else '',
                                'hometown': cols[4].get_text(strip=True) if len(cols) > 4 else '',
                                'height': cols[5].get_text(strip=True) if len(cols) > 5 else '',
                                'high_school_club': cols[6].get_text(strip=True) if len(cols) > 6 else ''
                            }
                            roster_data.append(athlete)

        # Método 2: Procurar por divs com dados de jogadores
        if not roster_data:
            print("🔍 Procurando em divs...")
            # Sidearm geralmente usa classes específicas
            player_classes = ['sidearm-roster-player', 'roster-card', 'player-card', 'athlete']
            for player_class in player_classes:
                players = soup.find_all('div', class_=player_class)
                if not players:
                    players = soup.find_all('article', class_=player_class)
                if players:
                    print(f"✅ Encontrados {len(players)} jogadores na classe '{player_class}'")
                    for player in players:
                        athlete = {
                            'school_name': school_name,
                            'url': url,
                            'jersey_number': '',
                            'name': '',
                            'position': '',
                            'class_year': '',
                            'hometown': '',
                            'height': '',
                            'high_school_club': ''
                        }

                        # Extrair número
                        number = player.find('span', class_='jersey-number')
                        if number:
                            athlete['jersey_number'] = number.get_text(strip=True)
                        else:
                            number = player.find('div', class_='number')
                            if number:
                                athlete['jersey_number'] = number.get_text(strip=True)

                        # Extrair nome
                        name = player.find('h3')
                        if name:
                            athlete['name'] = name.get_text(strip=True)
                        else:
                            name = player.find('div', class_='name')
                            if name:
                                athlete['name'] = name.get_text(strip=True)

                        # Extrair posição
                        pos = player.find('span', class_='position')
                        if pos:
                            athlete['position'] = pos.get_text(strip=True)

                        # Extrair outros detalhes
                        details = player.find_all('span', class_='detail')
                        for detail in details:
                            text = detail.get_text(strip=True)
                            if any(word in text.lower() for word in ['fr', 'so', 'jr', 'sr', 'freshman', 'sophomore', 'junior', 'senior', 'graduate']):
                                athlete['class_year'] = text
                            elif any(word in text.lower() for word in ['ft', 'in', '\'']):
                                athlete['height'] = text
                            elif any(word in text.lower() for word in ['hometown', 'city', 'town']):
                                athlete['hometown'] = text
                            elif any(word in text.lower() for word in ['high school', 'club', 'academy', 'prep']):
                                athlete['high_school_club'] = text

                        if athlete['name']:
                            roster_data.append(athlete)
                    break

        # Método 3: Procurar por qualquer lista de jogadores
        if not roster_data:
            print("🔍 Procurando em listas...")
            lists = soup.find_all(['ul', 'div'], class_=re.compile(r'roster|players|team'))
            for list_elem in lists:
                items = list_elem.find_all('li')
                if len(items) > 2:
                    for item in items:
                        text = item.get_text(strip=True)
                        if text and len(text) > 3:
                            athlete = {
                                'school_name': school_name,
                                'url': url,
                                'jersey_number': '',
                                'name': text.split('\n')[0] if '\n' in text else text[:50],
                                'position': '',
                                'class_year': '',
                                'hometown': '',
                                'height': '',
                                'high_school_club': ''
                            }
                            roster_data.append(athlete)
                    if roster_data:
                        break

        print(f"✅ Extraídos {len(roster_data)} atletas")

    except Exception as e:
        print(f"❌ Erro durante o scraping: {e}")
        import traceback
        traceback.print_exc()

    return pd.DataFrame(roster_data)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
import os

def append_to_roster_csv(new_df, filepath, key_columns=['school_name', 'name']):
    """Adiciona novos dados ao CSV evitando duplicatas."""
    os.makedirs(os.path.dirname(filepath), exist_ok=True)
    if os.path.exists(filepath):
        existing_df = pd.read_csv(filepath)
        print(f"📂 Arquivo existente: {len(existing_df)} registros")
        combined = pd.concat([existing_df, new_df], ignore_index=True)
        combined = combined.drop_duplicates(subset=key_columns, keep='last')
        print(f"✅ Adicionados {len(new_df)} registros, total {len(combined)}")
    else:
        combined = new_df
        print(f"📁 Criando novo arquivo com {len(combined)} registros")
    combined.to_csv(filepath, index=False, encoding='utf-8')
    print(f"💾 Salvo em: {filepath}")

def scrape_wmt_roster_requests(url, school_name):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36',
        'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9',
    }
    roster_data = []
    try:
        print(f"🌐 Acessando: {url}")
        response = requests.get(url, headers=headers, timeout=30)
        response.raise_for_status()
        soup = BeautifulSoup(response.content, 'html.parser')

        player_cards = []
        for selector in ['div.roster-card', 'div.roster__player', 'div.player-card']:
            cards = soup.select(selector)
            if cards:
                player_cards = cards
                print(f"✅ Encontrados {len(cards)} cards com seletor: {selector}")
                break

        if not player_cards:
            potential = soup.find_all('div', class_=re.compile(r'roster|player', re.I))
            for div in potential:
                if div.find(string=re.compile(r'#')):
                    player_cards.append(div)
            print(f"🔍 Encontrados {len(player_cards)} cards por fallback")

        for card in player_cards:
            athlete = {
                'school_name': school_name,
                'jersey_number': '',
                'name': '',
                'position': '',
                'class_year': '',
                'height': '',
                'hometown': '',
                'high_school_club': ''
            }

            number_elem = card.find(string=re.compile(r'^#'))
            if number_elem:
                num_text = number_elem.get_text(strip=True) if hasattr(number_elem, 'get_text') else number_elem
                match = re.search(r'#(\d+)', num_text)
                if match:
                    athlete['jersey_number'] = match.group(1)

            name_elem = card.find(['h3', 'h4'], class_=re.compile(r'name|title', re.I))
            if not name_elem:
                name_elem = card.find(class_=re.compile(r'player-name|full-name', re.I))
            if name_elem:
                raw_name = name_elem.get_text(strip=True)
                athlete['name'] = re.sub(r'Season\s+\d{4}-\d{2}', '', raw_name).strip()

            pos_elem = card.find(class_=re.compile(r'position', re.I))
            if pos_elem:
                athlete['position'] = pos_elem.get_text(strip=True)

            card_text = card.get_text(separator='\n', strip=True)
            lines = [line.strip() for line in card_text.split('\n') if line.strip()]
            lines = [line for line in lines if line != athlete['name'] and line != athlete['position']]

            for line in lines:
                if not athlete['height'] and re.search(r'\d+[\′\'"]\s*\d*[\′\'"]?|\d+-\d+', line):
                    athlete['height'] = line
                elif not athlete['class_year'] and re.search(r'\b(Fr\.?|So\.?|Jr\.?|Sr\.?|Gr\.?|Freshman|Sophomore|Junior|Senior|Graduate|Redshirt)\b', line, re.I):
                    athlete['class_year'] = line

            # PADRÃO ATUALIZADO: Suporta estados com ponto (ex: N.Y. ou Va.)
            hometown_pattern = r'[A-Z][a-zA-Z\s\.\-]+\s*,\s*([A-Z]{2}|[A-Z]\.[A-Z]\.|[A-Z][a-z]{1,2}\.)|[A-Z][a-zA-Z\s\.\-]+\s*,\s*[A-Z][a-z]+'

            for line in lines:
                if '/' in line:
                    parts = line.split('/', 1)
                    left = parts[0].strip()
                    right = parts[1].strip() if len(parts) > 1 else ''
                    if re.search(hometown_pattern, left):
                        athlete['hometown'] = left
                        athlete['high_school_club'] = right
                        break

            if not athlete['hometown']:
                for line in lines:
                    if re.search(hometown_pattern, line):
                        athlete['hometown'] = line
                        break

            if not athlete['high_school_club']:
                club_keywords = ['academy', 'school', 'club', 'fc', 'sc', 'united', 'breakers', 'solar', 'phoenix', 'atlanta', 'austin', 'dc united', 'right to dream', 'met oval', 'ifa', 'eintracht', 'frankfurt', 'cedar stars', 'morris united', 'new england revolution', 'atlanta united', 'westlake', 'grorud', 'fremad amager', 'afc dunstable', 'sta morris']
                for line in lines:
                    if any(k in line.lower() for k in club_keywords):
                        if line != athlete['hometown']:
                            athlete['high_school_club'] = line
                            break

            for key in ['hometown', 'high_school_club']:
                if athlete[key]:
                    athlete[key] = re.sub(r'Season\s+\d{4}-\d{2}', '', athlete[key]).strip()

            if athlete['name'] and (athlete['jersey_number'] or athlete['position']):
                roster_data.append(athlete)

        print(f"\n✅ Total extraído: {len(roster_data)} atletas")
    except Exception as e:
        print(f"❌ Erro: {e}")

    return pd.DataFrame(roster_data)

if __name__ == "__main__":
    test_url = "https://gostanford.com/sports/womens-soccer/roster"
    test_school = "Stanford University"
    df_new = scrape_wmt_roster_requests(test_url, test_school)

    if not df_new.empty:
        output_path = '/content/drive/MyDrive/RA_Project_EconDistance/raw_data/soccer_rosters.csv'
        append_to_roster_csv(df_new, output_path)
        display(df_new[['name', 'jersey_number', 'position', 'class_year', 'height', 'hometown', 'high_school_club']].head(15))
    else:
        print("❌ Nenhum dado extraído.")
```

```python
# Célula: Scraper PrestoSports usando requests (HTML estático)
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
import os

def append_to_roster_csv(new_df, filepath, key_columns=['school_name', 'name']):
    """Adiciona novos dados ao CSV evitando duplicatas."""
    os.makedirs(os.path.dirname(filepath), exist_ok=True)
    if os.path.exists(filepath):
        existing_df = pd.read_csv(filepath)
        print(f"📂 Arquivo existente: {len(existing_df)} registros")
        combined = pd.concat([existing_df, new_df], ignore_index=True)
        combined = combined.drop_duplicates(subset=key_columns, keep='last')
        print(f"✅ Adicionados {len(new_df)} registros, total {len(combined)}")
    else:
        combined = new_df
        print(f"📁 Criando novo arquivo com {len(combined)} registros")
    combined.to_csv(filepath, index=False, encoding='utf-8')
    print(f"💾 Salvo em: {filepath}")

def scrape_prestosports_roster(url, school_name):
    """
    Extrai dados do roster de sites PrestoSports.
    A página contém os dados em uma estrutura de lista.
    """
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36',
        'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9',
    }
    roster_data = []

    try:
        print(f"🌐 Acessando: {url}")
        response = requests.get(url, headers=headers, timeout=30)
        response.raise_for_status()
        soup = BeautifulSoup(response.content, 'html.parser')

        # Encontrar a lista de jogadores
        players_container = soup.find('ul', class_='sidearm-roster-players')
        if not players_container:
            print("❌ Não foi possível encontrar o container de jogadores.")
            return pd.DataFrame()

        players = players_container.find_all('li', class_='sidearm-roster-player')
        print(f"✅ Encontrados {len(players)} jogadores")

        for player in players:
            athlete = {
                'school_name': school_name,
                'jersey_number': '',
                'name': '',
                'position': '',
                'class_year': '',
                'height': '',
                'hometown': '',
                'high_school_club': ''
            }

            # Número da camisa
            number_elem = player.find('span', class_='sidearm-roster-player-jersey-number')
            if number_elem:
                athlete['jersey_number'] = number_elem.get_text(strip=True)

            # Nome (dentro de h3 > a)
            name_elem = player.find('h3')
            if name_elem:
                name_link = name_elem.find('a')
                if name_link:
                    athlete['name'] = name_link.get_text(strip=True)
                else:
                    athlete['name'] = name_elem.get_text(strip=True)

            # Posição (dentro de .sidearm-roster-player-position .text-bold)
            pos_elem = player.find('span', class_='text-bold')
            if pos_elem:
                athlete['position'] = pos_elem.get_text(strip=True)

            # Altura
            height_elem = player.find('span', class_='sidearm-roster-player-height')
            if height_elem:
                athlete['height'] = height_elem.get_text(strip=True)

            # Ano acadêmico (class year)
            # Pode estar em .sidearm-roster-player-academic-year
            year_elem = player.find('span', class_='sidearm-roster-player-academic-year')
            if year_elem:
                athlete['class_year'] = year_elem.get_text(strip=True)
            else:
                # Fallback: tentar capturar de dentro do .sidearm-roster-player-class-hometown
                class_div = player.find('div', class_='sidearm-roster-player-class-hometown')
                if class_div:
                    year_span = class_div.find('span', class_='sidearm-roster-player-academic-year')
                    if year_span:
                        athlete['class_year'] = year_span.get_text(strip=True)

            # Hometown e High School/Club
            # Eles estão dentro de .sidearm-roster-player-class-hometown
            # Mas podem estar separados ou juntos com "/"
            class_div = player.find('div', class_='sidearm-roster-player-class-hometown')
            if class_div:
                # Pode ter dois spans: um para hometown e outro para high school
                hometown_elem = class_div.find('span', class_='sidearm-roster-player-hometown')
                hs_elem = class_div.find('span', class_='sidearm-roster-player-highschool')

                if hometown_elem:
                    athlete['hometown'] = hometown_elem.get_text(strip=True)
                if hs_elem:
                    athlete['high_school_club'] = hs_elem.get_text(strip=True)

                # Se não separados, tentar o texto completo e dividir por "/"
                if not athlete['hometown'] and not athlete['high_school_club']:
                    full_text = class_div.get_text(separator=' ', strip=True)
                    if '/' in full_text:
                        parts = full_text.split('/', 1)
                        athlete['hometown'] = parts[0].strip()
                        athlete['high_school_club'] = parts[1].strip() if len(parts) > 1 else ''
                    else:
                        # Se não tem "/", provavelmente é só hometown
                        athlete['hometown'] = full_text
            else:
                # Fallback: procurar spans diretamente no player
                hometown_elem = player.find('span', class_='sidearm-roster-player-hometown')
                if hometown_elem:
                    athlete['hometown'] = hometown_elem.get_text(strip=True)
                hs_elem = player.find('span', class_='sidearm-roster-player-highschool')
                if hs_elem:
                    athlete['high_school_club'] = hs_elem.get_text(strip=True)

            # Limpeza: remover caracteres indesejados
            for key in athlete:
                if athlete[key]:
                    athlete[key] = re.sub(r'\s+', ' ', athlete[key]).strip()

            if athlete['name']:
                roster_data.append(athlete)
                print(f"  ✓ {athlete['name']} (#{athlete['jersey_number']}) - {athlete['position']} - {athlete['class_year']}")

        print(f"\n✅ Total extraído: {len(roster_data)} atletas")

    except Exception as e:
        print(f"❌ Erro: {e}")
        import traceback
        traceback.print_exc()

    return pd.DataFrame(roster_data)


# --- Teste com Denison University ---
test_url = "https://denisonbigred.com/sports/mens-soccer/roster"
test_school = "Denison University"

print("🚀 Testando scraper PrestoSports...")
print(f"📌 URL: {test_url}")
print(f"🏫 Escola: {test_school}")
print("-" * 60)

df_new = scrape_prestosports_roster(test_url, test_school)

if not df_new.empty:
    # Reordenar colunas para consistência
    colunas = ['school_name', 'jersey_number', 'name', 'position', 'class_year', 'height', 'hometown', 'high_school_club']
    df_new = df_new[colunas]

    # Mostrar primeiras linhas
    print("\n📊 Amostra dos dados extraídos:")
    print(df_new.head(10))

    # Salvar usando a função de append
    output_path = '/content/drive/MyDrive/RA_Project_EconDistance/raw_data/soccer_rosters.csv'
    append_to_roster_csv(df_new, output_path)

    print("\n✅ Dados adicionados ao arquivo principal!")
else:
    print("❌ Nenhum dado extraído.")
```

```python
def scrape_sidearm_roster(url, school_name, gender):
    """Sidearm: tenta requests, depois Selenium."""
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36'}
    try:
        print(f"  🌐 Sidearm: {url}")
        response = requests.get(url, headers=headers, timeout=20)
        if response.status_code != 200:
            raise Exception("Status não 200")
        soup = BeautifulSoup(response.content, 'html.parser')
        df = _parse_sidearm_soup(soup, school_name, gender)
        if df is not None and not df.empty:
            return df
        else:
            raise Exception("Nenhum dado encontrado")
    except Exception as e:
        print(f"    ⚠️ Requests falhou: {e}. Tentando Selenium...")
        return scrape_sidearm_selenium(url, school_name, gender)

def scrape_sidearm_selenium(url, school_name, gender):
    try:
        driver = setup_driver()
        driver.get(url)
        time.sleep(4)
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        driver.quit()
        return _parse_sidearm_soup(soup, school_name, gender)
    except Exception as e:
        print(f"    ❌ Erro no Selenium: {e}")
        return pd.DataFrame()

def _parse_sidearm_soup(soup, school_name, gender):
    roster_data = []
    # 1. Lista de jogadores (estilo Michigan)
    ul = soup.find('ul', class_='sidearm-roster-players')
    if ul:
        players = ul.find_all('li', class_='sidearm-roster-player')
        for p in players:
            athlete = {
                'school_name': school_name,
                'gender': gender,
                'jersey_number': '',
                'name': '',
                'position': '',
                'class_year': '',
                'height': '',
                'hometown': '',
                'high_school_club': ''
            }
            num = p.find('span', class_='sidearm-roster-player-jersey-number')
            if num:
                athlete['jersey_number'] = num.get_text(strip=True)
            name_tag = p.find('h3')
            if name_tag:
                link = name_tag.find('a')
                athlete['name'] = link.get_text(strip=True) if link else name_tag.get_text(strip=True)
            pos = p.find('span', class_='text-bold')
            if pos:
                athlete['position'] = pos.get_text(strip=True)
            height = p.find('span', class_='sidearm-roster-player-height')
            if height:
                athlete['height'] = height.get_text(strip=True)
            year = p.find('span', class_='sidearm-roster-player-academic-year')
            if year:
                athlete['class_year'] = year.get_text(strip=True)
            hometown_span = p.find('span', class_='sidearm-roster-player-hometown')
            if hometown_span:
                athlete['hometown'] = hometown_span.get_text(strip=True)
            hs_span = p.find('span', class_='sidearm-roster-player-highschool')
            if hs_span:
                athlete['high_school_club'] = hs_span.get_text(strip=True)
            if athlete['name']:
                roster_data.append(athlete)
    # 2. Tabela (fallback)
    if not roster_data:
        tables = soup.find_all('table')
        for table in tables:
            if 'roster' in str(table).lower():
                rows = table.find_all('tr')[1:]
                for row in rows:
                    cols = row.find_all('td')
                    if len(cols) >= 4:
                        athlete = {
                            'school_name': school_name,
                            'gender': gender,
                            'jersey_number': cols[0].get_text(strip=True),
                            'name': cols[1].get_text(strip=True),
                            'position': cols[2].get_text(strip=True),
                            'class_year': '',
                            'height': '',
                            'hometown': '',
                            'high_school_club': ''
                        }
                        third = cols[3].get_text(strip=True)
                        if re.search(r'\d+[\u2032\'\"]', third):
                            athlete['height'] = third
                        else:
                            athlete['class_year'] = third
                        if len(cols) > 4:
                            athlete['hometown'] = cols[4].get_text(strip=True)
                        if len(cols) > 5:
                            athlete['high_school_club'] = cols[5].get_text(strip=True)
                        roster_data.append(athlete)
    # 3. Cards genéricos (último fallback)
    if not roster_data:
        cards = soup.find_all('div', class_=re.compile(r'roster-card|player-card', re.I))
        for card in cards:
            athlete = {
                'school_name': school_name,
                'gender': gender,
                'jersey_number': '',
                'name': '',
                'position': '',
                'class_year': '',
                'height': '',
                'hometown': '',
                'high_school_club': ''
            }
            num = card.find(class_=re.compile(r'number|jersey', re.I))
            if num:
                athlete['jersey_number'] = num.get_text(strip=True).replace('#', '')
            name = card.find(['h3', 'h4'])
            if name:
                athlete['name'] = name.get_text(strip=True)
            pos = card.find(class_=re.compile(r'position', re.I))
            if pos:
                athlete['position'] = pos.get_text(strip=True)
            details = card.find_all(class_=re.compile(r'detail', re.I))
            for d in details:
                text = d.get_text(strip=True)
                if re.search(r'\d+[\u2032\'\"]', text):
                    athlete['height'] = text
                elif re.search(r'\b(Fr|So|Jr|Sr|Gr|Freshman|Sophomore|Junior|Senior|Graduate|Redshirt)\b', text, re.I):
                    athlete['class_year'] = text
                elif '/' in text:
                    parts = text.split('/', 1)
                    athlete['hometown'] = parts[0].strip()
                    athlete['high_school_club'] = parts[1].strip() if len(parts) > 1 else ''
                elif re.search(r'[A-Z][a-z]+\s*,\s*[A-Z]{2}\.?', text):
                    athlete['hometown'] = text
            if athlete['name']:
                roster_data.append(athlete)

    if roster_data:
        print(f"    ✓ Extraídos {len(roster_data)} atletas (Sidearm)")
    return pd.DataFrame(roster_data)
```

```python
def scrape_wmt_roster(url, school_name, gender):
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36'}
    try:
        print(f"  î0 WMT: {url}")
        response = requests.get(url, headers=headers, timeout=20)
        if response.status_code != 200:
            raise Exception("Status não 200")
        soup = BeautifulSoup(response.content, 'html.parser')
        df = _parse_wmt_soup(soup, school_name, gender)
        if df is not None and not df.empty:
            return df
        else:
            raise Exception("Nenhum dado encontrado")
    except Exception as e:
        print(f"    ⚠️ Requests falhou: {e}. Tentando Selenium...")
        return scrape_wmt_selenium(url, school_name, gender)

def scrape_wmt_selenium(url, school_name, gender):
    try:
        driver = setup_driver()
        driver.get(url)
        time.sleep(4)
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        driver.quit()
        return _parse_wmt_soup(soup, school_name, gender)
    except Exception as e:
        print(f"    ❌ Erro no Selenium: {e}")
        return pd.DataFrame()

def _parse_wmt_soup(soup, school_name, gender):
    roster_data = []

    # 1. TENTAR TABELA (LSU, etc.)
    table = soup.find('table', class_=re.compile(r'roster', re.I))
    if table:
        rows = table.find_all('tr')
        header_row = rows[0] if rows else None
        headers_text = []
        if header_row:
            headers_text = [th.get_text(strip=True).lower() for th in header_row.find_all(['th', 'td'])]
        for row in rows[1:]:
            cols = row.find_all('td')
            if len(cols) < 4:
                continue
            athlete = {
                'school_name': school_name,
                'gender': gender,
                'jersey_number': '',
                'name': '',
                'position': '',
                'class_year': '',
                'height': '',
                'hometown': '',
                'high_school_club': ''
            }
            for i, col in enumerate(cols):
                text = col.get_text(strip=True)
                if i < len(headers_text):
                    header = headers_text[i]
                    if 'no' in header or '#' in header:
                        athlete['jersey_number'] = text
                    elif 'name' in header:
                        athlete['name'] = text
                    elif 'pos' in header:
                        athlete['position'] = text
                    elif 'year' in header or 'class' in header:
                        athlete['class_year'] = text
                    elif 'height' in header:
                        athlete['height'] = text
                    elif 'hometown' in header:
                        athlete['hometown'] = text
                    elif 'high school' in header or 'school' in header:
                        athlete['high_school_club'] = text
                else:
                    if i == 0:
                        athlete['jersey_number'] = text
                    elif i == 1:
                        athlete['name'] = text
                    elif i == 2:
                        athlete['position'] = text
                    elif i == 3:
                        athlete['class_year'] = text
                    elif i == 4:
                        athlete['height'] = text
                    elif i == 5:
                        athlete['hometown'] = text
                    elif i == 6:
                        athlete['high_school_club'] = text
            if athlete['name']:
                roster_data.append(athlete)

    # 2. SE NÃO, TENTAR CARDS (Virginia, etc.)
    if not roster_data:
        # Tabela genérica (sem classe específica)
        player_cards = []
        for selector in ['div.roster-card', 'div.roster__player', 'div.player-card']:
            cards = soup.select(selector)
            if cards:
                player_cards = cards
                break
        if not player_cards:
            potential = soup.find_all('div', class_=re.compile(r'roster|player', re.I))
            for div in potential:
                if div.find(string=re.compile(r'#')):
                    player_cards.append(div)

        hometown_pattern = r'[A-Z][a-zA-Z\s\.\-]+\s*,\s*([A-Z]{2}|[A-Z]\.[A-Z]\.|[A-Z][a-z]{1,2}\.)|[A-Z][a-zA-Z\s\.\-]+\s*,\s*[A-Z][a-z]+'
        for card in player_cards:
            athlete = {
                'school_name': school_name,
                'gender': gender,
                'jersey_number': '',
                'name': '',
                'position': '',
                'class_year': '',
                'height': '',
                'hometown': '',
                'high_school_club': ''
            }
            number_elem = card.find(string=re.compile(r'^#'))
            if number_elem:
                match = re.search(r'#(\d+)', number_elem)
                if match:
                    athlete['jersey_number'] = match.group(1)
            name_elem = card.find(['h3', 'h4'], class_=re.compile(r'name|title', re.I))
            if name_elem:
                athlete['name'] = re.sub(r'Season\s+\d{4}-\d{2}', '', name_elem.get_text(strip=True)).strip()
            pos_elem = card.find(class_=re.compile(r'position', re.I))
            if pos_elem:
                athlete['position'] = pos.get_text(strip=True)
            card_text = card.get_text(separator='\n', strip=True)
            lines = [line.strip() for line in card_text.split('\n') if line.strip()]
            lines = [line for line in lines if line != athlete['name'] and line != athlete['position']]
            for line in lines:
                if not athlete['height'] and re.search(r'\d+[\u2032\'\"]\s*\d*[\u2032\'\"]?|\d+-\d+', line):
                    athlete['height'] = line
                elif not athlete['class_year'] and re.search(r'\b(Fr\.?|So\.?|Jr\.?|Sr\.?|Gr\.?|Freshman|Sophomore|Junior|Senior|Graduate|Redshirt)\b', line, re.I):
                    athlete['class_year'] = line
            for line in lines:
                if '/' in line:
                    parts = line.split('/', 1)
                    left = parts[0].strip()
                    right = parts[1].strip() if len(parts) > 1 else ''
                    if re.search(hometown_pattern, left):
                        athlete['hometown'] = left
                        athlete['high_school_club'] = right
                        break
            if not athlete['hometown']:
                for line in lines:
                    if re.search(hometown_pattern, line):
                        athlete['hometown'] = line
                        break
            if not athlete['high_school_club']:
                club_keywords = ['academy', 'school', 'club', 'fc', 'sc', 'united', 'breakers', 'solar', 'phoenix',
                                 'atlanta', 'austin', 'dc united', 'right to dream', 'met oval', 'ifa', 'eintracht',
                                 'frankfurt', 'cedar stars', 'morris united', 'new england revolution', 'atlanta united',
                                 'westlake', 'grorud', 'fremad amager', 'afc dunstable', 'sta morris']
                for line in lines:
                    if any(k in line.lower() for k in club_keywords):
                        if line != athlete['hometown']:
                            athlete['high_school_club'] = line
                            break
            if athlete['name']:
                roster_data.append(athlete)

    if roster_data:
        print(f"    ✓ Extraídos {len(roster_data)} atletas (WMT)")
    return pd.DataFrame(roster_data)
```

```python
# Célula: Scraper WMT Digital - processa todas as escolas da plataforma WMT
import pandas as pd
import requests
from bs4 import BeautifulSoup
import re
import time
import os
from selenium import webdriver
from selenium.webdriver.chrome.options import Options

# ------------------------------------------------------------------
# CONFIGURAÇÕES GLOBAIS
# ------------------------------------------------------------------
BASE_PATH = '/content/drive/MyDrive/RA_Project_EconDistance/raw_data/'
CSV_FILE = os.path.join(BASE_PATH, 'school_website_platforms.csv')
SUCCESS_FILE = os.path.join(BASE_PATH, 'soccer_rosters.csv')
FAILURE_FILE = os.path.join(BASE_PATH, 'soccer_roster_failures.csv')
DELAY = 3

# ------------------------------------------------------------------
# FUNÇÃO AUXILIAR: Append ao CSV
# ------------------------------------------------------------------
def append_to_roster_csv(new_df, filepath, key_columns=['school_name', 'name']):
    os.makedirs(os.path.dirname(filepath), exist_ok=True)
    if os.path.exists(filepath):
        existing_df = pd.read_csv(filepath)
        combined = pd.concat([existing_df, new_df], ignore_index=True)
        combined = combined.drop_duplicates(subset=key_columns, keep='last')
    else:
        combined = new_df
    combined.to_csv(filepath, index=False, encoding='utf-8')
    print(f"💾 Arquivo salvo em: {filepath}")

# ------------------------------------------------------------------
# SCRAPER WMT DIGITAL (versão requests, já validada)
# ------------------------------------------------------------------
def scrape_wmt_roster_requests(url, school_name):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36',
        'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9',
    }
    roster_data = []
    try:
        print(f"  🌐 WMT: {url}")
        response = requests.get(url, headers=headers, timeout=30)
        response.raise_for_status()
        soup = BeautifulSoup(response.content, 'html.parser')

        player_cards = []
        for selector in ['div.roster-card', 'div.roster__player', 'div.player-card']:
            cards = soup.select(selector)
            if cards:
                player_cards = cards
                print(f"    ✅ Encontrados {len(cards)} cards com seletor: {selector}")
                break

        if not player_cards:
            potential = soup.find_all('div', class_=re.compile(r'roster|player', re.I))
            for div in potential:
                if div.find(string=re.compile(r'#')):
                    player_cards.append(div)
            print(f"    🔍 Encontrados {len(player_cards)} cards por fallback")

        for card in player_cards:
            athlete = {
                'school_name': school_name,
                'jersey_number': '',
                'name': '',
                'position': '',
                'class_year': '',
                'height': '',
                'hometown': '',
                'high_school_club': ''
            }

            number_elem = card.find(string=re.compile(r'^#'))
            if number_elem:
                num_text = number_elem.get_text(strip=True) if hasattr(number_elem, 'get_text') else number_elem
                match = re.search(r'#(\d+)', num_text)
                if match:
                    athlete['jersey_number'] = match.group(1)

            name_elem = card.find(['h3', 'h4'], class_=re.compile(r'name|title', re.I))
            if not name_elem:
                name_elem = card.find(class_=re.compile(r'player-name|full-name', re.I))
            if name_elem:
                raw_name = name_elem.get_text(strip=True)
                athlete['name'] = re.sub(r'Season\s+\d{4}-\d{2}', '', raw_name).strip()

            pos_elem = card.find(class_=re.compile(r'position', re.I))
            if pos_elem:
                athlete['position'] = pos_elem.get_text(strip=True)

            card_text = card.get_text(separator='\n', strip=True)
            lines = [line.strip() for line in card_text.split('\n') if line.strip()]
            lines = [line for line in lines if line != athlete['name'] and line != athlete['position']]

            for line in lines:
                if not athlete['height'] and re.search(r'\d+[\′\'"]\s*\d*[\′\'"]?|\d+-\d+', line):
                    athlete['height'] = line
                elif not athlete['class_year'] and re.search(r'\b(Fr\.?|So\.?|Jr\.?|Sr\.?|Gr\.?|Freshman|Sophomore|Junior|Senior|Graduate|Redshirt)\b', line, re.I):
                    athlete['class_year'] = line

            # Padrão para capturar hometown com estados abreviados (ex: N.Y., Va.)
            hometown_pattern = r'[A-Z][a-zA-Z\s\.\-]+\s*,\s*([A-Z]{2}|[A-Z]\.[A-Z]\.|[A-Z][a-z]{1,2}\.)|[A-Z][a-zA-Z\s\.\-]+\s*,\s*[A-Z][a-z]+'

            for line in lines:
                if '/' in line:
                    parts = line.split('/', 1)
                    left = parts[0].strip()
                    right = parts[1].strip() if len(parts) > 1 else ''
                    if re.search(hometown_pattern, left):
                        athlete['hometown'] = left
                        athlete['high_school_club'] = right
                        break

            if not athlete['hometown']:
                for line in lines:
                    if re.search(hometown_pattern, line):
                        athlete['hometown'] = line
                        break

            if not athlete['high_school_club']:
                club_keywords = [
                    'academy', 'school', 'club', 'fc', 'sc', 'united', 'breakers', 'solar', 'phoenix',
                    'atlanta', 'austin', 'dc united', 'right to dream', 'met oval', 'ifa', 'eintracht',
                    'frankfurt', 'cedar stars', 'morris united', 'new england revolution', 'atlanta united',
                    'westlake', 'grorud', 'fremad amager', 'afc dunstable', 'sta morris'
                ]
                for line in lines:
                    if any(k in line.lower() for k in club_keywords):
                        if line != athlete['hometown']:
                            athlete['high_school_club'] = line
                            break

            for key in ['hometown', 'high_school_club']:
                if athlete[key]:
                    athlete[key] = re.sub(r'Season\s+\d{4}-\d{2}', '', athlete[key]).strip()

            if athlete['name'] and (athlete['jersey_number'] or athlete['position']):
                roster_data.append(athlete)

        print(f"    ✓ Extraídos {len(roster_data)} atletas (WMT)")
        return pd.DataFrame(roster_data)

    except Exception as e:
        print(f"    ❌ Erro no WMT (requests): {e}")
        return pd.DataFrame()

# ------------------------------------------------------------------
# SCRAPER WMT DIGITAL COM SELENIUM (fallback)
# ------------------------------------------------------------------
def scrape_wmt_selenium(url, school_name):
    try:
        chrome_options = Options()
        chrome_options.add_argument('--headless')
        chrome_options.add_argument('--no-sandbox')
        chrome_options.add_argument('--disable-dev-shm-usage')
        chrome_options.add_argument('--window-size=1920,1080')
        chrome_options.add_argument('--user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36')
        driver = webdriver.Chrome(options=chrome_options)
        driver.get(url)
        time.sleep(4)
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        driver.quit()

        # Reutilizar a mesma lógica de extração (cards) em cima do HTML renderizado
        roster_data = []
        player_cards = []
        for selector in ['div.roster-card', 'div.roster__player', 'div.player-card']:
            cards = soup.select(selector)
            if cards:
                player_cards = cards
                break
        if not player_cards:
            potential = soup.find_all('div', class_=re.compile(r'roster|player', re.I))
            for div in potential:
                if div.find(string=re.compile(r'#')):
                    player_cards.append(div)

        for card in player_cards:
            athlete = {
                'school_name': school_name,
                'jersey_number': '',
                'name': '',
                'position': '',
                'class_year': '',
                'height': '',
                'hometown': '',
                'high_school_club': ''
            }
            number_elem = card.find(string=re.compile(r'^#'))
            if number_elem:
                match = re.search(r'#(\d+)', number_elem)
                if match:
                    athlete['jersey_number'] = match.group(1)
            name_elem = card.find(['h3', 'h4'], class_=re.compile(r'name|title', re.I))
            if name_elem:
                athlete['name'] = re.sub(r'Season\s+\d{4}-\d{2}', '', name_elem.get_text(strip=True)).strip()
            pos_elem = card.find(class_=re.compile(r'position', re.I))
            if pos_elem:
                athlete['position'] = pos_elem.get_text(strip=True)
            card_text = card.get_text(separator='\n', strip=True)
            lines = [line.strip() for line in card_text.split('\n') if line.strip()]
            lines = [line for line in lines if line != athlete['name'] and line != athlete['position']]
            for line in lines:
                if not athlete['height'] and re.search(r'\d+[\′\'"]\s*\d*[\′\'"]?|\d+-\d+', line):
                    athlete['height'] = line
                elif not athlete['class_year'] and re.search(r'\b(Fr\.?|So\.?|Jr\.?|Sr\.?|Gr\.?|Freshman|Sophomore|Junior|Senior|Graduate|Redshirt)\b', line, re.I):
                    athlete['class_year'] = line
            # Hometown e high school
            hometown_pattern = r'[A-Z][a-zA-Z\s\.\-]+\s*,\s*([A-Z]{2}|[A-Z]\.[A-Z]\.|[A-Z][a-z]{1,2}\.)|[A-Z][a-zA-Z\s\.\-]+\s*,\s*[A-Z][a-z]+'
            for line in lines:
                if '/' in line:
                    parts = line.split('/', 1)
                    left = parts[0].strip()
                    right = parts[1].strip() if len(parts) > 1 else ''
                    if re.search(hometown_pattern, left):
                        athlete['hometown'] = left
                        athlete['high_school_club'] = right
                        break
            if not athlete['hometown']:
                for line in lines:
                    if re.search(hometown_pattern, line):
                        athlete['hometown'] = line
                        break
            if not athlete['high_school_club']:
                club_keywords = ['academy', 'school', 'club', 'fc', 'sc', 'united', 'breakers', 'solar', 'phoenix',
                                 'atlanta', 'austin', 'dc united', 'right to dream', 'met oval', 'ifa', 'eintracht',
                                 'frankfurt', 'cedar stars', 'morris united', 'new england revolution', 'atlanta united',
                                 'westlake', 'grorud', 'fremad amager', 'afc dunstable', 'sta morris']
                for line in lines:
                    if any(k in line.lower() for k in club_keywords):
                        if line != athlete['hometown']:
                            athlete['high_school_club'] = line
                            break
            if athlete['name']:
                roster_data.append(athlete)
        print(f"    ✓ Extraídos {len(roster_data)} atletas (WMT via Selenium)")
        return pd.DataFrame(roster_data)
    except Exception as e:
        print(f"    ❌ Erro no WMT (Selenium): {e}")
        return pd.DataFrame()

# ------------------------------------------------------------------
# ORQUESTRADOR WMT
# ------------------------------------------------------------------
print("📂 Carregando arquivo de plataformas...")
df = pd.read_csv(CSV_FILE)

# Filtra apenas WMT Digital
df_wmt = df[df['platform'] == 'WMT Digital']
print(f"Escolas WMT a processar: {len(df_wmt)}")

failures = []

for idx, row in df_wmt.iterrows():
    school = row['school_name']
    gender = row.get('gender', '')
    roster_url = row['roster_url']

    if pd.isna(roster_url) or not str(roster_url).startswith('http'):
        print(f"[{idx+1}] {school} ({gender}) - Sem URL válida")
        failures.append({'school': school, 'gender': gender, 'reason': 'No roster URL', 'url': roster_url})
        continue

    print(f"\n[{idx+1}] {school} ({gender})")
    print(f"   URL: {roster_url}")

    # Tentar primeiro com requests (rápido)
    df_result = scrape_wmt_roster_requests(roster_url, school)

    # Se falhar ou não retornar dados, tentar Selenium
    if df_result.empty:
        print(f"   ⚠️ Requests não retornou dados. Tentando Selenium...")
        df_result = scrape_wmt_selenium(roster_url, school)

    if not df_result.empty:
        df_result['gender'] = gender
        append_to_roster_csv(df_result, SUCCESS_FILE)
        print(f"   ✅ Sucesso: {len(df_result)} atletas")
    else:
        print(f"   ⚠️ Nenhum dado retornado")
        failures.append({'school': school, 'gender': gender, 'reason': 'No data found', 'url': roster_url})

    time.sleep(DELAY)

# Salvar falhas, se houver
if failures:
    fail_df = pd.DataFrame(failures)
    if os.path.exists(FAILURE_FILE):
        fail_df.to_csv(FAILURE_FILE, mode='a', header=False, index=False, encoding='utf-8')
    else:
        fail_df.to_csv(FAILURE_FILE, index=False, encoding='utf-8')
    print(f"💾 Falhas registradas em {FAILURE_FILE}")

print("\n✅ Processamento WMT concluído.")
```

```python
# -*- coding: utf-8 -*-
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
import os
import time
from selenium import webdriver
from selenium.webdriver.chrome.options import Options

# ======================== CONFIGURAÇÕES ========================
BASE_PATH = '/content/drive/MyDrive/RA_Project_EconDistance/raw_data/'
CSV_FILE = os.path.join(BASE_PATH, 'school_website_platforms.csv')
SUCCESS_FILE = os.path.join(BASE_PATH, 'soccer_rosters.csv')
FAILURE_FILE = os.path.join(BASE_PATH, 'soccer_roster_failures.csv')

# ======================== FUNÇÕES AUXILIARES ========================
def setup_driver():
    """Configura o driver do Selenium para headless."""
    chrome_options = Options()
    chrome_options.add_argument('--headless')
    chrome_options.add_argument('--no-sandbox')
    chrome_options.add_argument('--disable-dev-shm-usage')
    chrome_options.add_argument('--window-size=1920,1080')
    chrome_options.add_argument('--user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36')
    return webdriver.Chrome(options=chrome_options)

def append_to_roster_csv(new_df, filepath, key_columns=['school_name', 'name']):
    """Adiciona novos dados ao CSV, evitando duplicatas."""
    os.makedirs(os.path.dirname(filepath), exist_ok=True)
    if os.path.exists(filepath):
        existing_df = pd.read_csv(filepath)
        combined = pd.concat([existing_df, new_df], ignore_index=True)
        combined = combined.drop_duplicates(subset=key_columns, keep='last')
    else:
        combined = new_df
    combined.to_csv(filepath, index=False, encoding='utf-8')
    print(f"💾 Arquivo salvo em: {filepath}")

# ======================== PARSER ESPECÍFICO PARA STANFORD ========================
def parse_stanford_cards(soup):
    """Extrai dados dos cards da Stanford (gostanford.com)."""
    roster_data = []
    cards = soup.select('div.roster-card-item')
    print(f"🔍 Encontrados {len(cards)} cards principais (Stanford)")

    # Palavras-chave para excluir staff
    STAFF_KEYWORDS = ['coach', 'director', 'assistant', 'manager', 'sports medicine',
                      'performance', 'nutrition', 'psychology', 'therapist', 'academic',
                      'equipment', 'communications', 'sports performance']

    height_pattern = r'(\d+[\′\'"]\s*\d*[\′\'"]?)'
    class_pattern = r'\b(Fr\.?|So\.?|Jr\.?|Sr\.?|Gr\.?|Freshman|Sophomore|Junior|Senior|Graduate|Redshirt)\b'
    hometown_pattern = r'[A-Z][a-zA-Z\s\.\-]+,\s*[A-Z]{2}\.?|[A-Z][a-zA-Z\s\.\-]+,\s*[A-Z][a-z]+'

    for card in cards:
        athlete = {
            'school_name': '',
            'jersey_number': '',
            'name': '',
            'position': '',
            'class_year': '',
            'height': '',
            'hometown': '',
            'high_school_club': ''
        }

        # Número
        num_elem = card.select_one('strong.roster-card-item__jersey-number')
        if num_elem:
            athlete['jersey_number'] = num_elem.get_text(strip=True)

        # Nome (prioridade: alt da imagem)
        img = card.select_one('img')
        if img and img.get('alt'):
            athlete['name'] = img['alt'].strip()
        else:
            name_elem = card.select_one('.roster-card-item__content strong:first-child')
            if name_elem:
                athlete['name'] = name_elem.get_text(strip=True)

        # Posição
        pos_elem = card.select_one('strong.roster-card-item__position')
        if pos_elem:
            athlete['position'] = pos_elem.get_text(strip=True)

        # Se a posição contiver palavras de staff, pula este card
        if athlete['position'] and any(k in athlete['position'].lower() for k in STAFF_KEYWORDS):
            print(f"  ⏭️ Pulando staff: {athlete['name']} - {athlete['position']}")
            continue

        # Se não tem nome ou posição, pula
        if not athlete['name'] or not athlete['position']:
            continue

        # Extrai os detalhes usando os rótulos
        card_text = card.get_text(separator='\n', strip=True)
        lines = [line.strip() for line in card_text.split('\n') if line.strip()]

        # Primeiro, procura linhas com rótulos explícitos
        for line in lines:
            lower = line.lower()
            if 'hometown:' in lower:
                athlete['hometown'] = line.split(':', 1)[-1].strip()
            elif 'high school:' in lower:
                athlete['high_school_club'] = line.split(':', 1)[-1].strip()
            elif 'height:' in lower:
                athlete['height'] = line.split(':', 1)[-1].strip()
            elif 'class:' in lower or 'year:' in lower:
                athlete['class_year'] = line.split(':', 1)[-1].strip()

        # Fallback: se ainda faltar algum campo, tenta com regex
        if not athlete['height'] or not athlete['class_year'] or not athlete['hometown']:
            for line in lines:
                if not athlete['height'] and re.search(height_pattern, line):
                    athlete['height'] = re.search(height_pattern, line).group(1)
                elif not athlete['class_year'] and re.search(class_pattern, line, re.I):
                    athlete['class_year'] = re.search(class_pattern, line, re.I).group(0)
                elif not athlete['hometown'] and re.search(hometown_pattern, line):
                    athlete['hometown'] = re.search(hometown_pattern, line).group(0)
                elif not athlete['high_school_club'] and any(k in line.lower() for k in ['high school', 'academy', 'club']):
                    athlete['high_school_club'] = line

        # Remove possíveis prefixos numéricos que possam ter escapado (ex: "9Jaden Thomas")
        if athlete['name']:
            athlete['name'] = re.sub(r'^\d+', '', athlete['name']).strip()

        roster_data.append(athlete)
        print(f"  ✓ {athlete['name']} (#{athlete['jersey_number']}) - {athlete['position']}")

    return roster_data

def scrape_stanford(url, school_name):
    """Função específica para Stanford (usa Selenium)."""
    print("🔍 Stanford detectado – usando Selenium...")
    driver = setup_driver()
    try:
        driver.get(url)
        time.sleep(5)  # aguarda renderização
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        roster_data = parse_stanford_cards(soup)
        for athlete in roster_data:
            athlete['school_name'] = school_name
        print(f"✅ Extraídos {len(roster_data)} atletas (Stanford)")
        return pd.DataFrame(roster_data)
    except Exception as e:
        print(f"❌ Erro no Selenium para Stanford: {e}")
        return pd.DataFrame()
    finally:
        driver.quit()

# ======================== PARSER GENÉRICO PARA WMT DIGITAL (VIRGINIA, ETC.) ========================
def scrape_generic_wmt(url, school_name):
    """Scraper genérico usando requests (funciona para Virginia e outros Sidearm)."""
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36',
        'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9',
    }
    try:
        print(f"🌐 Acessando (requests): {url}")
        response = requests.get(url, headers=headers, timeout=30)
        response.raise_for_status()
        soup = BeautifulSoup(response.content, 'html.parser')

        # ===== PROCURAR CARDS =====
        player_cards = []
        for selector in ['div.roster-card', 'div.roster__player', 'div.player-card', 'li.roster-item']:
            cards = soup.select(selector)
            if cards:
                player_cards = cards
                print(f"✅ Encontrados {len(cards)} cards com seletor: {selector}")
                break

        if not player_cards:
            potential = soup.find_all('div', class_=re.compile(r'roster|player', re.I))
            for div in potential:
                if div.find(string=re.compile(r'#')):
                    player_cards.append(div)
            print(f"🔍 Encontrados {len(player_cards)} cards por fallback")

        roster_data = []
        for card in player_cards:
            athlete = {
                'school_name': school_name,
                'jersey_number': '',
                'name': '',
                'position': '',
                'class_year': '',
                'height': '',
                'hometown': '',
                'high_school_club': ''
            }

            # --- Número ---
            number_elem = card.find(string=re.compile(r'^#'))
            if number_elem:
                num_text = number_elem.get_text(strip=True) if hasattr(number_elem, 'get_text') else number_elem
                match = re.search(r'#(\d+)', num_text)
                if match:
                    athlete['jersey_number'] = match.group(1)

            # --- Nome ---
            name_elem = card.find(['h3', 'h4'], class_=re.compile(r'name|title', re.I))
            if not name_elem:
                name_elem = card.find(class_=re.compile(r'player-name|full-name', re.I))
            if name_elem:
                raw_name = name_elem.get_text(strip=True)
                athlete['name'] = re.sub(r'Season\s+\d{4}-\d{2}', '', raw_name).strip()

            # --- Posição ---
            pos_elem = card.find(class_=re.compile(r'position', re.I))
            if pos_elem:
                athlete['position'] = pos_elem.get_text(strip=True)

            # --- Extração de detalhes usando rótulos e blocos estruturados ---
            # Tenta encontrar um bloco de detalhes (ex: <div class="details">)
            details_container = card.find(class_=re.compile(r'details|info|bio', re.I))
            if details_container:
                # Coleta todos os textos de elementos filhos que podem conter os rótulos
                detail_texts = [elem.get_text(strip=True) for elem in details_container.find_all(['li', 'div', 'p', 'span'])]
            else:
                # Se não houver bloco de detalhes, usa todo o texto do card
                detail_texts = [card.get_text(separator='\n', strip=True)]

            # Junta todos os textos em uma lista de linhas
            all_lines = []
            for text in detail_texts:
                all_lines.extend([line.strip() for line in text.split('\n') if line.strip()])

            # Processa cada linha em busca dos campos
            for line in all_lines:
                lower = line.lower()
                # Hometown
                if 'hometown:' in lower:
                    # Extrai após os dois pontos e remove espaços
                    athlete['hometown'] = line.split(':', 1)[-1].strip()
                # High School
                elif 'high school:' in lower:
                    athlete['high_school_club'] = line.split(':', 1)[-1].strip()
                # Height
                elif 'height:' in lower:
                    athlete['height'] = line.split(':', 1)[-1].strip()
                # Class/Year
                elif 'class:' in lower or 'year:' in lower:
                    athlete['class_year'] = line.split(':', 1)[-1].strip()
                # Caso não tenha rótulo explícito, tenta capturar por regex
                else:
                    # Altura (padrão como 5'6", 6'2", etc.)
                    if not athlete['height'] and re.search(r'\d+[\′\'"]\s*\d*[\′\'"]?', line):
                        athlete['height'] = re.search(r'\d+[\′\'"]\s*\d*[\′\'"]?', line).group(0)
                    # Ano acadêmico (Freshman, Sophomore, Junior, Senior, Redshirt, etc.)
                    elif not athlete['class_year'] and re.search(r'\b(Fr\.?|So\.?|Jr\.?|Sr\.?|Gr\.?|Freshman|Sophomore|Junior|Senior|Graduate|Redshirt)\b', line, re.I):
                        athlete['class_year'] = re.search(r'\b(Fr\.?|So\.?|Jr\.?|Sr\.?|Gr\.?|Freshman|Sophomore|Junior|Senior|Graduate|Redshirt)\b', line, re.I).group(0)
                    # Hometown (cidade, estado)
                    elif not athlete['hometown'] and re.search(r'[A-Z][a-zA-Z\s\.\-]+,\s*[A-Z]{2}', line):
                        athlete['hometown'] = re.search(r'[A-Z][a-zA-Z\s\.\-]+,\s*[A-Z]{2}', line).group(0)
                    # High School / Club (palavras-chave)
                    elif not athlete['high_school_club'] and any(k in lower for k in ['high school', 'academy', 'club', 'fc', 'sc', 'united', 'breakers', 'solar']):
                        athlete['high_school_club'] = line

            # Se ainda não encontrou class_year, tenta por um padrão mais amplo
            if not athlete['class_year']:
                # Exemplo: "Freshman" pode estar em texto sem rótulo
                for line in all_lines:
                    match = re.search(r'\b(Fr\.?|So\.?|Jr\.?|Sr\.?|Gr\.?|Freshman|Sophomore|Junior|Senior|Graduate|Redshirt)\b', line, re.I)
                    if match:
                        athlete['class_year'] = match.group(0)
                        break

            # Se ainda não encontrou height, tenta por padrão numérico com polegadas
            if not athlete['height']:
                for line in all_lines:
                    match = re.search(r'\d+[\′\'"]\s*\d*[\′\'"]?', line)
                    if match:
                        athlete['height'] = match.group(0)
                        break

            # Limpeza dos campos: remover rótulos caso tenham escapado
            for field in ['hometown', 'high_school_club', 'class_year', 'height']:
                if athlete[field]:
                    # Remove qualquer rótulo que possa ter ficado (ex: "Hometown: " no início)
                    athlete[field] = re.sub(r'^(hometown|high school|class|year|height)\s*:\s*', '', athlete[field], flags=re.I).strip()

            if athlete['name'] and (athlete['jersey_number'] or athlete['position']):
                roster_data.append(athlete)

        print(f"✅ Total extraído: {len(roster_data)} atletas")
        return pd.DataFrame(roster_data)

    except Exception as e:
        print(f"❌ Erro no requests: {e}")
        return pd.DataFrame()

# ======================== ORQUESTRADOR PRINCIPAL ========================
def scrape_wmt_roster(url, school_name):
    """Função principal que decide qual scraper usar baseado na URL."""
    if 'gostanford.com' in url:
        return scrape_stanford(url, school_name)
    else:
        return scrape_generic_wmt(url, school_name)

# ======================== EXECUÇÃO ========================
if __name__ == "__main__":
    # Instalação do chromedriver no Colab (necessário apenas no Colab)
    try:
        import google.colab
        print("📦 Ambiente Colab detectado. Instalando chromedriver...")
        !apt-get update
        !apt install chromium-chromedriver
        !cp /usr/lib/chromium-browser/chromedriver /usr/bin
        !pip install selenium
    except ImportError:
        pass  # não está no Colab, assume que o chromedriver já está no PATH

    # Carrega planilha com as escolas
    print("📂 Carregando arquivo de plataformas...")
    if not os.path.exists(CSV_FILE):
        print(f"❌ Arquivo não encontrado: {CSV_FILE}")
        exit(1)

    df = pd.read_csv(CSV_FILE)
    # Filtra apenas escolas com plataforma WMT Digital (ou ajuste conforme sua coluna)
    if 'platform' in df.columns:
        df_wmt = df[df['platform'] == 'WMT Digital']
    else:
        print("⚠️ Coluna 'platform' não encontrada. Processando todas as linhas.")
        df_wmt = df

    print(f"Escolas a processar: {len(df_wmt)}")
    failures = []

    for idx, row in df_wmt.iterrows():
        school = row['school_name']
        gender = row.get('gender', '')
        roster_url = row.get('roster_url', '')

        if pd.isna(roster_url) or not str(roster_url).startswith('http'):
            print(f"\n[{idx+1}] {school} ({gender}) - Sem URL válida")
            failures.append({'school': school, 'gender': gender, 'reason': 'No roster URL', 'url': roster_url})
            continue

        print(f"\n[{idx+1}] {school} ({gender})")
        print(f"   URL: {roster_url}")

        df_result = scrape_wmt_roster(roster_url, school)

        if df_result is not None and not df_result.empty:
            df_result['gender'] = gender
            append_to_roster_csv(df_result, SUCCESS_FILE)
            print(f"   ✅ Sucesso: {len(df_result)} atletas")
        else:
            print(f"   ⚠️ Nenhum dado retornado")
            failures.append({'school': school, 'gender': gender, 'reason': 'No data found', 'url': roster_url})

        time.sleep(2)  # pausa entre requisições

    # Salva falhas
    if failures:
        fail_df = pd.DataFrame(failures)
        if os.path.exists(FAILURE_FILE):
            fail_df.to_csv(FAILURE_FILE, mode='a', header=False, index=False, encoding='utf-8')
        else:
            fail_df.to_csv(FAILURE_FILE, index=False, encoding='utf-8')
        print(f"💾 Falhas registradas em {FAILURE_FILE}")
```

```python
# -*- coding: utf-8 -*-
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
import os
import time
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# ======================== CONFIGURAÇÕES ========================
BASE_PATH = '/content/drive/MyDrive/RA_Project_EconDistance/raw_data/'
CSV_FILE = os.path.join(BASE_PATH, 'school_website_platforms.csv')
SUCCESS_FILE = os.path.join(BASE_PATH, 'soccer_rosters.csv')
FAILURE_FILE = os.path.join(BASE_PATH, 'soccer_roster_failures.csv')

# ======================== FUNÇÕES AUXILIARES ========================
def setup_driver():
    """Configura o driver do Selenium para headless."""
    chrome_options = Options()
    chrome_options.add_argument('--headless')
    chrome_options.add_argument('--no-sandbox')
    chrome_options.add_argument('--disable-dev-shm-usage')
    chrome_options.add_argument('--window-size=1920,1080')
    chrome_options.add_argument('--user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36')
    return webdriver.Chrome(options=chrome_options)

def append_to_roster_csv(new_df, filepath, key_columns=['school_name', 'name']):
    """Adiciona novos dados ao CSV, evitando duplicatas."""
    os.makedirs(os.path.dirname(filepath), exist_ok=True)
    # Reordena colunas para consistência
    column_order = ['school_name', 'gender', 'jersey_number', 'name', 'position', 'class_year', 'height', 'hometown', 'high_school_club']
    new_df = new_df[[col for col in column_order if col in new_df.columns]]
    if os.path.exists(filepath):
        existing_df = pd.read_csv(filepath)
        combined = pd.concat([existing_df, new_df], ignore_index=True)
        combined = combined.drop_duplicates(subset=key_columns, keep='last')
    else:
        combined = new_df
    combined.to_csv(filepath, index=False, encoding='utf-8')
    print(f"💾 Arquivo salvo em: {filepath}")

def split_hometown_school(text):
    """Divide uma string que contém hometown e high school separados por '/'."""
    if '/' not in text:
        return text, ''
    parts = text.split('/', 1)
    left = parts[0].strip()
    right = parts[1].strip() if len(parts) > 1 else ''
    # Verifica se a parte esquerda parece uma cidade/estado (padrão: Cidade, Estado)
    if re.search(r'[A-Z][a-zA-Z\s\.\-]+,\s*[A-Z]{2}', left) or re.search(r'[A-Z][a-zA-Z\s\.\-]+,\s*[A-Z][a-z]+', left):
        return left, right
    return text, ''

def extract_compound_line(line):
    """
    Tenta extrair campos de uma linha composta como:
    "6′1″185 lbsSophomoreAustin FCHometownAustin, TexasHigh SchoolWestlake High School"
    Retorna dicionário com campos encontrados.
    """
    result = {'height': '', 'class_year': '', 'hometown': '', 'high_school_club': ''}
    temp = line

    # 1. Altura
    height_match = re.search(r'(\d+[\′\'"]\s*\d*[\′\'"]?)', temp)
    if height_match:
        result['height'] = height_match.group(1)
        temp = temp.replace(height_match.group(1), '')

    # 2. Ano acadêmico (inclui Redshirt, Graduate, etc.)
    year_match = re.search(r'\b(Redshirt\s+)?(Fr\.?|So\.?|Jr\.?|Sr\.?|Gr\.?|Freshman|Sophomore|Junior|Senior|Graduate|Redshirt|Redshirt Freshman|Redshirt Sophomore|Redshirt Junior|Redshirt Senior|Graduate Student)\b', temp, re.I)
    if year_match:
        result['class_year'] = year_match.group(0).strip()
        temp = temp.replace(year_match.group(0), '')

    # 3. Hometown (após "Hometown")
    hometown_match = re.search(r'Hometown\s*([A-Z][a-zA-Z\s\.\-]+,\s*[A-Z]{2})', temp, re.I)
    if not hometown_match:
        hometown_match = re.search(r'Hometown\s*([A-Z][a-zA-Z\s\.\-]+,\s*[A-Z][a-z]+)', temp, re.I)
    if hometown_match:
        result['hometown'] = hometown_match.group(1).strip()
        temp = temp.replace(hometown_match.group(0), '')

    # 4. High School (após "High School" ou "Previous School")
    hs_match = re.search(r'High School\s*([A-Z][a-zA-Z\s\.\-]+)', temp, re.I)
    if not hs_match:
        hs_match = re.search(r'Previous School\s*([A-Z][a-zA-Z\s\.\-]+)', temp, re.I)
    if hs_match:
        result['high_school_club'] = hs_match.group(1).strip()
        temp = temp.replace(hs_match.group(0), '')

    # Se ainda não temos high school, tenta capturar o que sobrou (pode ser nome do clube)
    if not result['high_school_club'] and temp.strip():
        # Remove números e palavras-chave residuais
        temp_clean = re.sub(r'\d+', '', temp)
        temp_clean = re.sub(r'(lbs|kg|Hometown|High School|Previous School)', '', temp_clean, flags=re.I)
        if temp_clean.strip():
            result['high_school_club'] = temp_clean.strip()

    return result

# ======================== PARSER ESPECÍFICO PARA STANFORD ========================
def parse_stanford_cards(soup):
    """Extrai dados dos cards da Stanford (gostanford.com)."""
    roster_data = []
    cards = soup.select('div.roster-card-item')
    print(f"🔍 Encontrados {len(cards)} cards principais (Stanford)")

    STAFF_KEYWORDS = ['coach', 'director', 'assistant', 'manager', 'sports medicine',
                      'performance', 'nutrition', 'psychology', 'therapist', 'academic',
                      'equipment', 'communications', 'sports performance']

    height_pattern = r'(\d+[\′\'"]\s*\d*[\′\'"]?)'
    class_pattern = r'\b(Redshirt\s+)?(Fr\.?|So\.?|Jr\.?|Sr\.?|Gr\.?|Freshman|Sophomore|Junior|Senior|Graduate|Redshirt|Redshirt Freshman|Redshirt Sophomore|Redshirt Junior|Redshirt Senior|Graduate Student)\b'
    hometown_pattern = r'[A-Z][a-zA-Z\s\.\-]+,\s*[A-Z]{2}\.?|[A-Z][a-zA-Z\s\.\-]+,\s*[A-Z][a-z]+'

    for card in cards:
        athlete = {
            'school_name': '',
            'jersey_number': '',
            'name': '',
            'position': '',
            'class_year': '',
            'height': '',
            'hometown': '',
            'high_school_club': ''
        }

        # Número
        num_elem = card.select_one('strong.roster-card-item__jersey-number')
        if num_elem:
            athlete['jersey_number'] = num_elem.get_text(strip=True)

        # Nome (prioridade: alt da imagem)
        img = card.select_one('img')
        if img and img.get('alt'):
            athlete['name'] = img['alt'].strip()
        else:
            name_elem = card.select_one('.roster-card-item__content strong:first-child')
            if name_elem:
                athlete['name'] = name_elem.get_text(strip=True)

        # Posição
        pos_elem = card.select_one('strong.roster-card-item__position')
        if pos_elem:
            athlete['position'] = pos_elem.get_text(strip=True)

        # Pular staff
        if athlete['position'] and any(k in athlete['position'].lower() for k in STAFF_KEYWORDS):
            print(f"  ⏭️ Pulando staff: {athlete['name']} - {athlete['position']}")
            continue

        if not athlete['name'] or not athlete['position']:
            continue

        # Extração com rótulos
        card_text = card.get_text(separator='\n', strip=True)
        lines = [line.strip() for line in card_text.split('\n') if line.strip()]

        # Rótulos explícitos
        for line in lines:
            lower = line.lower()
            if 'hometown:' in lower:
                athlete['hometown'] = line.split(':', 1)[-1].strip()
            elif 'high school:' in lower:
                athlete['high_school_club'] = line.split(':', 1)[-1].strip()
            elif 'height:' in lower:
                athlete['height'] = line.split(':', 1)[-1].strip()
            elif 'class:' in lower or 'year:' in lower:
                athlete['class_year'] = line.split(':', 1)[-1].strip()

        # Fallback regex
        if not athlete['height'] or not athlete['class_year'] or not athlete['hometown']:
            for line in lines:
                if not athlete['height'] and re.search(height_pattern, line):
                    athlete['height'] = re.search(height_pattern, line).group(1)
                elif not athlete['class_year'] and re.search(class_pattern, line, re.I):
                    athlete['class_year'] = re.search(class_pattern, line, re.I).group(0)
                elif not athlete['hometown'] and re.search(hometown_pattern, line):
                    athlete['hometown'] = re.search(hometown_pattern, line).group(0)
                elif not athlete['high_school_club'] and any(k in line.lower() for k in ['high school', 'academy', 'club']):
                    athlete['high_school_club'] = line

        # Remove prefixos numéricos do nome
        if athlete['name']:
            athlete['name'] = re.sub(r'^\d+', '', athlete['name']).strip()

        roster_data.append(athlete)
        print(f"  ✓ {athlete['name']} (#{athlete['jersey_number']}) - {athlete['position']}")

    return roster_data

def scrape_stanford(url, school_name):
    """Função específica para Stanford (usa Selenium)."""
    print("🔍 Stanford detectado – usando Selenium...")
    driver = setup_driver()
    try:
        driver.get(url)
        time.sleep(5)  # aguarda renderização
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        roster_data = parse_stanford_cards(soup)
        for athlete in roster_data:
            athlete['school_name'] = school_name
        print(f"✅ Extraídos {len(roster_data)} atletas (Stanford)")
        return pd.DataFrame(roster_data)
    except Exception as e:
        print(f"❌ Erro no Selenium para Stanford: {e}")
        return pd.DataFrame()
    finally:
        driver.quit()

# ======================== PARSER GENÉRICO PARA WMT DIGITAL (VIRGINIA, ETC.) ========================
def scrape_generic_wmt(url, school_name):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36',
        'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9',
    }
    try:
        print(f"🌐 Acessando (requests): {url}")
        response = requests.get(url, headers=headers, timeout=30)
        response.raise_for_status()
        soup = BeautifulSoup(response.content, 'html.parser')

        # ----- LOCALIZAR CARDS -----
        player_cards = []
        for selector in ['div.roster__player', 'div.roster-card', 'div.player-card', 'li.roster-item']:
            cards = soup.select(selector)
            if cards:
                player_cards = cards
                print(f"✅ Encontrados {len(cards)} cards com seletor: {selector}")
                break

        if not player_cards:
            potential = soup.find_all('div', class_=re.compile(r'roster|player', re.I))
            for div in potential:
                if div.find(string=re.compile(r'#')):
                    player_cards.append(div)
            print(f"🔍 Encontrados {len(player_cards)} cards por fallback")

        roster_data = []
        for card in player_cards:
            athlete = {
                'school_name': school_name,
                'jersey_number': '',
                'name': '',
                'position': '',
                'class_year': '',
                'height': '',
                'hometown': '',
                'high_school_club': ''
            }

            # --- Número ---
            number_elem = card.find(string=re.compile(r'^#'))
            if number_elem:
                num_text = number_elem.get_text(strip=True) if hasattr(number_elem, 'get_text') else number_elem
                match = re.search(r'#(\d+)', num_text)
                if match:
                    athlete['jersey_number'] = match.group(1)

            # --- Nome ---
            name_elem = card.find(['h3', 'h4'], class_=re.compile(r'name|title', re.I))
            if not name_elem:
                name_elem = card.find(class_=re.compile(r'player-name|full-name', re.I))
            if name_elem:
                raw_name = name_elem.get_text(strip=True)
                athlete['name'] = re.sub(r'Season\s+\d{4}-\d{2}', '', raw_name).strip()

            # --- Posição ---
            pos_elem = card.find(class_=re.compile(r'position', re.I))
            if pos_elem:
                athlete['position'] = pos_elem.get_text(strip=True)

            # ----- BUSCA DIRETA POR ELEMENTOS COM CLASSES ESPECÍFICAS -----
            # Tenta encontrar hometown e high school em elementos com classes comuns
            hometown_elem = card.find(class_=re.compile(r'hometown|city', re.I))
            if hometown_elem:
                athlete['hometown'] = hometown_elem.get_text(strip=True)
            highschool_elem = card.find(class_=re.compile(r'highschool|high-school|prev-school|previous-school', re.I))
            if highschool_elem:
                athlete['high_school_club'] = highschool_elem.get_text(strip=True)

            # ----- EXTRAÇÃO ESTRUTURADA (PRIORITÁRIA) -----
            if not athlete['hometown'] or not athlete['high_school_club']:
                details_container = card.find(class_=re.compile(r'details|info|bio', re.I))
                if details_container:
                    detail_items = details_container.find_all(['li', 'div', 'p'])
                    for item in detail_items:
                        text = item.get_text(strip=True)
                        lower = text.lower()
                        if 'hometown:' in lower:
                            athlete['hometown'] = text.split(':', 1)[-1].strip()
                        elif 'high school:' in lower:
                            athlete['high_school_club'] = text.split(':', 1)[-1].strip()
                        elif 'height:' in lower:
                            athlete['height'] = text.split(':', 1)[-1].strip()
                        elif 'class:' in lower or 'year:' in lower:
                            athlete['class_year'] = text.split(':', 1)[-1].strip()
                        elif '/' in text and not athlete['hometown'] and not athlete['high_school_club']:
                            parts = text.split('/', 1)
                            left = parts[0].strip()
                            right = parts[1].strip() if len(parts) > 1 else ''
                            if re.search(r'[A-Z][a-zA-Z\s\.\-]+,\s*[A-Z]{2}', left) or re.search(r'[A-Z][a-zA-Z\s\.\-]+,\s*[A-Z][a-z]+', left):
                                athlete['hometown'] = left
                                athlete['high_school_club'] = right
                        elif not athlete['height'] and re.search(r'\d+[\′\'"]\s*\d*[\′\'"]?', text):
                            athlete['height'] = re.search(r'\d+[\′\'"]\s*\d*[\′\'"]?', text).group(0)
                        elif not athlete['class_year'] and re.search(r'\b(Fr\.?|So\.?|Jr\.?|Sr\.?|Gr\.?|Freshman|Sophomore|Junior|Senior|Graduate|Redshirt)\b', text, re.I):
                            athlete['class_year'] = re.search(r'\b(Fr\.?|So\.?|Jr\.?|Sr\.?|Gr\.?|Freshman|Sophomore|Junior|Senior|Graduate|Redshirt)\b', text, re.I).group(0)

            # ----- FALLBACK: TEXTO COMPLETO DO CARD -----
            if (not athlete['hometown'] or not athlete['high_school_club'] or not athlete['height'] or not athlete['class_year']):
                card_text = card.get_text(separator='\n', strip=True)
                lines = [line.strip() for line in card_text.split('\n') if line.strip()]
                lines = [line for line in lines if line != athlete['name'] and line != athlete['position']]

                for line in lines:
                    lower = line.lower()
                    # 1. Rótulos explícitos
                    if 'hometown:' in lower:
                        athlete['hometown'] = line.split(':', 1)[-1].strip()
                    elif 'high school:' in lower:
                        athlete['high_school_club'] = line.split(':', 1)[-1].strip()
                    elif 'height:' in lower:
                        athlete['height'] = line.split(':', 1)[-1].strip()
                    elif 'class:' in lower or 'year:' in lower:
                        athlete['class_year'] = line.split(':', 1)[-1].strip()
                    # 2. Separação por barra
                    elif '/' in line and not athlete['hometown'] and not athlete['high_school_club']:
                        parts = line.split('/', 1)
                        left = parts[0].strip()
                        right = parts[1].strip() if len(parts) > 1 else ''
                        if re.search(r'[A-Z][a-zA-Z\s\.\-]+,\s*[A-Z]{2}', left) or re.search(r'[A-Z][a-zA-Z\s\.\-]+,\s*[A-Z][a-z]+', left):
                            athlete['hometown'] = left
                            athlete['high_school_club'] = right
                    # 3. Linhas compostas (vários campos concatenados)
                    else:
                        # Detecta se a linha contém indicadores de múltiplos campos
                        if any(k in line for k in ['lbs', 'Hometown', 'High School', 'Previous School']):
                            # Extração sequencial sem lookbehind
                            temp = line
                            # Altura
                            height_match = re.search(r'(\d+[\′\'"]\s*\d*[\′\'"]?)', temp)
                            if height_match and not athlete['height']:
                                athlete['height'] = height_match.group(1)
                                temp = temp.replace(height_match.group(1), '')
                            # Ano acadêmico
                            year_match = re.search(r'\b(Fr\.?|So\.?|Jr\.?|Sr\.?|Gr\.?|Freshman|Sophomore|Junior|Senior|Graduate|Redshirt)\b', temp, re.I)
                            if year_match and not athlete['class_year']:
                                athlete['class_year'] = year_match.group(0)
                                temp = temp.replace(year_match.group(0), '')
                            # Hometown (após "Hometown" - aceita estado com 2 letras ou nome completo)
                            hometown_match = re.search(r'Hometown\s*([A-Z][a-zA-Z\s\.\-]+,\s*[A-Z]{2})', temp, re.I)
                            if not hometown_match:
                                hometown_match = re.search(r'Hometown\s*([A-Z][a-zA-Z\s\.\-]+,\s*[A-Z][a-z]+)', temp, re.I)
                            if hometown_match and not athlete['hometown']:
                                athlete['hometown'] = hometown_match.group(1).strip()
                                temp = temp.replace(hometown_match.group(0), '')
                            # High School (após "High School")
                            hs_match = re.search(r'High School\s*([A-Z][a-zA-Z\s\.\-]+)', temp, re.I)
                            if hs_match and not athlete['high_school_club']:
                                athlete['high_school_club'] = hs_match.group(1).strip()
                                temp = temp.replace(hs_match.group(0), '')
                            # Se ainda não temos clube, tenta capturar o que sobrou entre o ano e o hometown
                            if not athlete['high_school_club']:
                                # Remove números e palavras-chave para tentar isolar o nome do clube
                                temp_clean = re.sub(r'\d+[\′\'"]\s*\d*[\′\'"]?', '', temp)
                                temp_clean = re.sub(r'\b(Fr\.?|So\.?|Jr\.?|Sr\.?|Gr\.?|Freshman|Sophomore|Junior|Senior|Graduate|Redshirt)\b', '', temp_clean, flags=re.I)
                                temp_clean = re.sub(r'Hometown\s*[A-Z][a-zA-Z\s\.\-]+,\s*[A-Z]{2}', '', temp_clean, flags=re.I)
                                temp_clean = re.sub(r'Hometown\s*[A-Z][a-zA-Z\s\.\-]+,\s*[A-Z][a-z]+', '', temp_clean, flags=re.I)
                                # O que sobrar pode ser o clube
                                if temp_clean.strip() and not athlete['high_school_club']:
                                    athlete['high_school_club'] = temp_clean.strip()
                        else:
                            # Fallback padrão: regex para altura, ano, hometown, high school
                            if not athlete['height'] and re.search(r'\d+[\′\'"]\s*\d*[\′\'"]?', line):
                                athlete['height'] = re.search(r'\d+[\′\'"]\s*\d*[\′\'"]?', line).group(0)
                            if not athlete['class_year'] and re.search(r'\b(Fr\.?|So\.?|Jr\.?|Sr\.?|Gr\.?|Freshman|Sophomore|Junior|Senior|Graduate|Redshirt)\b', line, re.I):
                                athlete['class_year'] = re.search(r'\b(Fr\.?|So\.?|Jr\.?|Sr\.?|Gr\.?|Freshman|Sophomore|Junior|Senior|Graduate|Redshirt)\b', line, re.I).group(0)
                            if not athlete['hometown'] and re.search(r'[A-Z][a-zA-Z\s\.\-]+,\s*[A-Z]{2}', line):
                                athlete['hometown'] = re.search(r'[A-Z][a-zA-Z\s\.\-]+,\s*[A-Z]{2}', line).group(0)
                            if not athlete['hometown'] and re.search(r'[A-Z][a-zA-Z\s\.\-]+,\s*[A-Z][a-z]+', line):
                                athlete['hometown'] = re.search(r'[A-Z][a-zA-Z\s\.\-]+,\s*[A-Z][a-z]+', line).group(0)
                            if not athlete['high_school_club'] and any(k in lower for k in ['high school', 'academy', 'club', 'fc', 'sc', 'united', 'breakers', 'solar']):
                                athlete['high_school_club'] = line

            # Limpeza final: remove rótulos residuais e o prefixo "Hometown"
            for field in ['hometown', 'high_school_club', 'class_year', 'height']:
                if athlete[field]:
                    # Remove rótulos como "Hometown:", "High School:", etc.
                    athlete[field] = re.sub(r'^(hometown|high school|class|year|height)\s*:\s*', '', athlete[field], flags=re.I).strip()
                    # Remove a palavra "Hometown" sozinha no início
                    if field == 'hometown' and athlete[field].lower().startswith('hometown'):
                        athlete[field] = athlete[field][8:].strip()
                    # Remove "Previous School" se aparecer
                    if field == 'high_school_club' and 'previous school' in athlete[field].lower():
                        athlete[field] = re.sub(r'previous school\s*', '', athlete[field], flags=re.I).strip()

            if athlete['name']:
                roster_data.append(athlete)

        print(f"✅ Total extraído: {len(roster_data)} atletas")
        return pd.DataFrame(roster_data)

    except Exception as e:
        print(f"❌ Erro no requests: {e}")
        return pd.DataFrame()

def extract_wmt_from_soup(soup, url):
    """Função auxiliar que extrai dados do BeautifulSoup."""
    players = []
    # Tenta seletores comuns
    player_cards = []
    for selector in ['div.roster__player', 'div.roster-card', 'div.player-card', 'li.roster-item']:
        cards = soup.select(selector)
        if cards:
            player_cards = cards
            print(f"✅ Encontrados {len(cards)} cards com seletor: {selector}")
            break

    if not player_cards:
        potential = soup.find_all('div', class_=re.compile(r'roster|player', re.I))
        for div in potential:
            if div.find(string=re.compile(r'#')):
                player_cards.append(div)
        print(f"🔍 Encontrados {len(player_cards)} cards por fallback")

    for card in player_cards:
        athlete = {
            'school_name': '',
            'jersey_number': '',
            'name': '',
            'position': '',
            'class_year': '',
            'height': '',
            'hometown': '',
            'high_school_club': ''
        }

        # --- Número ---
        number_elem = card.find(string=re.compile(r'^#'))
        if number_elem:
            num_text = number_elem.get_text(strip=True) if hasattr(number_elem, 'get_text') else number_elem
            match = re.search(r'#(\d+)', num_text)
            if match:
                athlete['jersey_number'] = match.group(1)

        # --- Nome ---
        name_elem = card.find(['h3', 'h4'], class_=re.compile(r'name|title', re.I))
        if not name_elem:
            name_elem = card.find(class_=re.compile(r'player-name|full-name', re.I))
        if name_elem:
            raw_name = name_elem.get_text(strip=True)
            athlete['name'] = re.sub(r'Season\s+\d{4}-\d{2}', '', raw_name).strip()

        # --- Posição ---
        pos_elem = card.find(class_=re.compile(r'position', re.I))
        if pos_elem:
            athlete['position'] = pos_elem.get_text(strip=True)

        # --- Extração de detalhes ---
        # Tenta encontrar bloco de detalhes estruturado
        details_container = card.find(class_=re.compile(r'details|info|bio', re.I))
        if details_container:
            detail_texts = [elem.get_text(strip=True) for elem in details_container.find_all(['li', 'div', 'p', 'span'])]
        else:
            detail_texts = [card.get_text(separator='\n', strip=True)]

        all_lines = []
        for text in detail_texts:
            all_lines.extend([line.strip() for line in text.split('\n') if line.strip()])

        # Processa cada linha
        for line in all_lines:
            lower = line.lower()
            # Rótulos explícitos
            if 'hometown:' in lower:
                athlete['hometown'] = line.split(':', 1)[-1].strip()
            elif 'high school:' in lower:
                athlete['high_school_club'] = line.split(':', 1)[-1].strip()
            elif 'height:' in lower:
                athlete['height'] = line.split(':', 1)[-1].strip()
            elif 'class:' in lower or 'year:' in lower:
                athlete['class_year'] = line.split(':', 1)[-1].strip()
            # Separação por barra
            elif '/' in line and not athlete['hometown'] and not athlete['high_school_club']:
                hometown, school = split_hometown_school(line)
                if hometown != line:
                    athlete['hometown'] = hometown
                    athlete['high_school_club'] = school
            # Linhas compostas (ex: com múltiplos campos)
            elif any(k in line for k in ['lbs', 'Hometown', 'High School', 'Previous School']):
                compound = extract_compound_line(line)
                if compound['height']:
                    athlete['height'] = compound['height']
                if compound['class_year']:
                    athlete['class_year'] = compound['class_year']
                if compound['hometown']:
                    athlete['hometown'] = compound['hometown']
                if compound['high_school_club']:
                    athlete['high_school_club'] = compound['high_school_club']
            # Fallback regex
            else:
                if not athlete['height'] and re.search(r'\d+[\′\'"]\s*\d*[\′\'"]?', line):
                    athlete['height'] = re.search(r'\d+[\′\'"]\s*\d*[\′\'"]?', line).group(0)
                if not athlete['class_year'] and re.search(r'\b(Redshirt\s+)?(Fr\.?|So\.?|Jr\.?|Sr\.?|Gr\.?|Freshman|Sophomore|Junior|Senior|Graduate|Redshirt|Redshirt Freshman|Redshirt Sophomore|Redshirt Junior|Redshirt Senior|Graduate Student)\b', line, re.I):
                    athlete['class_year'] = re.search(r'\b(Redshirt\s+)?(Fr\.?|So\.?|Jr\.?|Sr\.?|Gr\.?|Freshman|Sophomore|Junior|Senior|Graduate|Redshirt|Redshirt Freshman|Redshirt Sophomore|Redshirt Junior|Redshirt Senior|Graduate Student)\b', line, re.I).group(0)
                if not athlete['hometown'] and re.search(r'[A-Z][a-zA-Z\s\.\-]+,\s*[A-Z]{2}', line):
                    athlete['hometown'] = re.search(r'[A-Z][a-zA-Z\s\.\-]+,\s*[A-Z]{2}', line).group(0)
                if not athlete['high_school_club'] and any(k in lower for k in ['high school', 'academy', 'club', 'fc', 'sc', 'united', 'breakers', 'solar']):
                    athlete['high_school_club'] = line

        # Limpeza final
        for field in ['hometown', 'high_school_club', 'class_year', 'height']:
            if athlete[field]:
                athlete[field] = re.sub(r'^(hometown|high school|class|year|height)\s*:\s*', '', athlete[field], flags=re.I).strip()
                # Remove "Hometown" sozinho no início
                if field == 'hometown' and athlete[field].lower().startswith('hometown'):
                    athlete[field] = athlete[field][8:].strip()
                if field == 'high_school_club' and athlete[field].lower().startswith('high school'):
                    athlete[field] = athlete[field][11:].strip()

        if athlete['name'] and (athlete['jersey_number'] or athlete['position']):
            players.append(athlete)

    return pd.DataFrame(players) if players else None

# ======================== ORQUESTRADOR PRINCIPAL ========================
def scrape_wmt_roster(url, school_name):
    """Função principal que decide qual scraper usar baseado na URL."""
    if 'gostanford.com' in url:
        return scrape_stanford(url, school_name)
    else:
        return scrape_generic_wmt(url, school_name)

# ======================== EXECUÇÃO ========================
if __name__ == "__main__":
    # Instalação do chromedriver no Colab (necessário apenas no Colab)
    try:
        import google.colab
        print("📦 Ambiente Colab detectado. Instalando chromedriver...")
        !apt-get update
        !apt install chromium-chromedriver
        !cp /usr/lib/chromium-browser/chromedriver /usr/bin
        !pip install selenium
    except ImportError:
        pass  # não está no Colab, assume que o chromedriver já está no PATH

    # Carrega planilha com as escolas
    print("📂 Carregando arquivo de plataformas...")
    if not os.path.exists(CSV_FILE):
        print(f"❌ Arquivo não encontrado: {CSV_FILE}")
        exit(1)

    df = pd.read_csv(CSV_FILE)
    # Filtra apenas escolas com plataforma WMT Digital (ou ajuste conforme sua coluna)
    if 'platform' in df.columns:
        df_wmt = df[df['platform'] == 'WMT Digital']
    else:
        print("⚠️ Coluna 'platform' não encontrada. Processando todas as linhas.")
        df_wmt = df

    print(f"Escolas a processar: {len(df_wmt)}")
    failures = []

    for idx, row in df_wmt.iterrows():
        school = row['school_name']
        gender = row.get('gender', '')
        roster_url = row.get('roster_url', '')

        if pd.isna(roster_url) or not str(roster_url).startswith('http'):
            print(f"\n[{idx+1}] {school} ({gender}) - Sem URL válida")
            failures.append({'school': school, 'gender': gender, 'reason': 'No roster URL', 'url': roster_url})
            continue

        print(f"\n[{idx+1}] {school} ({gender})")
        print(f"   URL: {roster_url}")

        df_result = scrape_wmt_roster(roster_url, school)

        if df_result is not None and not df_result.empty:
            df_result['gender'] = gender
            append_to_roster_csv(df_result, SUCCESS_FILE)
            print(f"   ✅ Sucesso: {len(df_result)} atletas")
        else:
            print(f"   ⚠️ Nenhum dado retornado")
            failures.append({'school': school, 'gender': gender, 'reason': 'No data found', 'url': roster_url})

        time.sleep(2)  # pausa entre requisições

    # Salva falhas
    if failures:
        fail_df = pd.DataFrame(failures)
        if os.path.exists(FAILURE_FILE):
            fail_df.to_csv(FAILURE_FILE, mode='a', header=False, index=False, encoding='utf-8')
        else:
            fail_df.to_csv(FAILURE_FILE, index=False, encoding='utf-8')
        print(f"💾 Falhas registradas em {FAILURE_FILE}")
```

```python
# -*- coding: utf-8 -*-
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
import os
import time
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# ======================== CONFIGURAÇÕES ========================
BASE_PATH = '/content/drive/MyDrive/RA_Project_EconDistance/raw_data/'
CSV_FILE = os.path.join(BASE_PATH, 'school_website_platforms.csv')
SUCCESS_FILE = os.path.join(BASE_PATH, 'soccer_rosters.csv')
FAILURE_FILE = os.path.join(BASE_PATH, 'soccer_roster_failures.csv')

# ======================== FUNÇÕES AUXILIARES ========================
def setup_driver():
    chrome_options = Options()
    chrome_options.add_argument('--headless')
    chrome_options.add_argument('--no-sandbox')
    chrome_options.add_argument('--disable-dev-shm-usage')
    chrome_options.add_argument('--window-size=1920,1080')
    chrome_options.add_argument('--user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36')
    return webdriver.Chrome(options=chrome_options)

def append_to_roster_csv(new_df, filepath, key_columns=['school_name', 'name']):
    os.makedirs(os.path.dirname(filepath), exist_ok=True)
    column_order = ['school_name', 'gender', 'jersey_number', 'name', 'position', 'class_year', 'height', 'hometown', 'high_school_club']
    new_df = new_df[[col for col in column_order if col in new_df.columns]]
    if os.path.exists(filepath):
        existing_df = pd.read_csv(filepath)
        combined = pd.concat([existing_df, new_df], ignore_index=True)
        combined = combined.drop_duplicates(subset=key_columns, keep='last')
    else:
        combined = new_df
    combined.to_csv(filepath, index=False, encoding='utf-8')
    print(f"💾 Arquivo salvo em: {filepath}")

def split_hometown_school(text):
    """Divide string com hometown e high school separados por '/'."""
    if '/' not in text:
        return text, ''
    parts = text.split('/', 1)
    left = parts[0].strip()
    right = parts[1].strip() if len(parts) > 1 else ''
    if re.search(r'[A-Z][a-zA-Z\s\.\-]+,\s*[A-Z]{2}', left) or re.search(r'[A-Z][a-zA-Z\s\.\-]+,\s*[A-Z][a-z]+', left):
        return left, right
    return text, ''

def extract_compound_line(line):
    """
    Extrai campos de linhas compostas como:
    "6′1″185 lbsSophomoreAustin FCHometownAustin, TexasHigh SchoolWestlake High School"
    """
    result = {'height': '', 'class_year': '', 'hometown': '', 'high_school_club': ''}
    temp = line

    # Altura
    height_match = re.search(r'(\d+[\′\'"]\s*\d*[\′\'"]?)', temp)
    if height_match:
        result['height'] = height_match.group(1)
        temp = temp.replace(height_match.group(1), '')

    # Ano acadêmico
    year_match = re.search(r'\b(Redshirt\s+)?(Fr\.?|So\.?|Jr\.?|Sr\.?|Gr\.?|Freshman|Sophomore|Junior|Senior|Graduate|Redshirt|Redshirt Freshman|Redshirt Sophomore|Redshirt Junior|Redshirt Senior|Graduate Student)\b', temp, re.I)
    if year_match:
        result['class_year'] = year_match.group(0).strip()
        temp = temp.replace(year_match.group(0), '')

    # Hometown (após "Hometown")
    hometown_match = re.search(r'Hometown\s*([A-Z][a-zA-Z\s\.\-]+,\s*[A-Z]{2})', temp, re.I)
    if not hometown_match:
        hometown_match = re.search(r'Hometown\s*([A-Z][a-zA-Z\s\.\-]+,\s*[A-Z][a-z]+)', temp, re.I)
    if hometown_match:
        result['hometown'] = hometown_match.group(1).strip()
        temp = temp.replace(hometown_match.group(0), '')

    # High School (após "High School" ou "Previous School")
    hs_match = re.search(r'High School\s*([A-Z][a-zA-Z\s\.\-]+)', temp, re.I)
    if not hs_match:
        hs_match = re.search(r'Previous School\s*([A-Z][a-zA-Z\s\.\-]+)', temp, re.I)
    if hs_match:
        result['high_school_club'] = hs_match.group(1).strip()
        temp = temp.replace(hs_match.group(0), '')

    # Último recurso: pegar o que sobrou como clube
    if not result['high_school_club'] and temp.strip():
        temp_clean = re.sub(r'\d+', '', temp)
        temp_clean = re.sub(r'(lbs|kg|Hometown|High School|Previous School)', '', temp_clean, flags=re.I)
        if temp_clean.strip():
            result['high_school_club'] = temp_clean.strip()

    return result

# ======================== PARSER ESPECÍFICO PARA STANFORD ========================
def parse_stanford_cards(soup):
    roster_data = []
    cards = soup.select('div.roster-card-item')
    print(f"🔍 Encontrados {len(cards)} cards principais (Stanford)")

    STAFF_KEYWORDS = ['coach', 'director', 'assistant', 'manager', 'sports medicine',
                      'performance', 'nutrition', 'psychology', 'therapist', 'academic',
                      'equipment', 'communications', 'sports performance']

    height_pattern = r'(\d+[\′\'"]\s*\d*[\′\'"]?)'
    class_pattern = r'\b(Redshirt\s+)?(Fr\.?|So\.?|Jr\.?|Sr\.?|Gr\.?|Freshman|Sophomore|Junior|Senior|Graduate|Redshirt|Redshirt Freshman|Redshirt Sophomore|Redshirt Junior|Redshirt Senior|Graduate Student)\b'
    hometown_pattern = r'[A-Z][a-zA-Z\s\.\-]+,\s*[A-Z]{2}\.?|[A-Z][a-zA-Z\s\.\-]+,\s*[A-Z][a-z]+'

    for card in cards:
        athlete = {
            'school_name': '',
            'jersey_number': '',
            'name': '',
            'position': '',
            'class_year': '',
            'height': '',
            'hometown': '',
            'high_school_club': ''
        }

        num_elem = card.select_one('strong.roster-card-item__jersey-number')
        if num_elem:
            athlete['jersey_number'] = num_elem.get_text(strip=True)

        img = card.select_one('img')
        if img and img.get('alt'):
            athlete['name'] = img['alt'].strip()
        else:
            name_elem = card.select_one('.roster-card-item__content strong:first-child')
            if name_elem:
                athlete['name'] = name_elem.get_text(strip=True)

        pos_elem = card.select_one('strong.roster-card-item__position')
        if pos_elem:
            athlete['position'] = pos_elem.get_text(strip=True)

        if athlete['position'] and any(k in athlete['position'].lower() for k in STAFF_KEYWORDS):
            print(f"  ⏭️ Pulando staff: {athlete['name']} - {athlete['position']}")
            continue

        if not athlete['name'] or not athlete['position']:
            continue

        card_text = card.get_text(separator='\n', strip=True)
        lines = [line.strip() for line in card_text.split('\n') if line.strip()]

        for line in lines:
            lower = line.lower()
            if 'hometown:' in lower:
                athlete['hometown'] = line.split(':', 1)[-1].strip()
            elif 'high school:' in lower:
                athlete['high_school_club'] = line.split(':', 1)[-1].strip()
            elif 'height:' in lower:
                athlete['height'] = line.split(':', 1)[-1].strip()
            elif 'class:' in lower or 'year:' in lower:
                athlete['class_year'] = line.split(':', 1)[-1].strip()

        if not athlete['height'] or not athlete['class_year'] or not athlete['hometown']:
            for line in lines:
                if not athlete['height'] and re.search(height_pattern, line):
                    athlete['height'] = re.search(height_pattern, line).group(1)
                elif not athlete['class_year'] and re.search(class_pattern, line, re.I):
                    athlete['class_year'] = re.search(class_pattern, line, re.I).group(0)
                elif not athlete['hometown'] and re.search(hometown_pattern, line):
                    athlete['hometown'] = re.search(hometown_pattern, line).group(0)
                elif not athlete['high_school_club'] and any(k in line.lower() for k in ['high school', 'academy', 'club']):
                    athlete['high_school_club'] = line

        if athlete['name']:
            athlete['name'] = re.sub(r'^\d+', '', athlete['name']).strip()

        roster_data.append(athlete)
        print(f"  ✓ {athlete['name']} (#{athlete['jersey_number']}) - {athlete['position']}")

    return roster_data

def scrape_stanford(url, school_name):
    print("🔍 Stanford detectado – usando Selenium...")
    driver = setup_driver()
    try:
        driver.get(url)
        time.sleep(5)
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        roster_data = parse_stanford_cards(soup)
        for athlete in roster_data:
            athlete['school_name'] = school_name
        print(f"✅ Extraídos {len(roster_data)} atletas (Stanford)")
        return pd.DataFrame(roster_data)
    except Exception as e:
        print(f"❌ Erro no Selenium para Stanford: {e}")
        return pd.DataFrame()
    finally:
        driver.quit()

# ======================== PARSER GENÉRICO PARA WMT DIGITAL ========================
def scrape_generic_wmt(url, school_name):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36',
        'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9',
    }
    try:
        print(f"🌐 Acessando (requests): {url}")
        response = requests.get(url, headers=headers, timeout=30)
        response.raise_for_status()
        soup = BeautifulSoup(response.content, 'html.parser')

        # ----- LOCALIZAR CARDS -----
        player_cards = []
        for selector in ['div.roster__player', 'div.roster-card', 'div.player-card', 'li.roster-item']:
            cards = soup.select(selector)
            if cards:
                player_cards = cards
                print(f"✅ Encontrados {len(cards)} cards com seletor: {selector}")
                break

        if not player_cards:
            potential = soup.find_all('div', class_=re.compile(r'roster|player', re.I))
            for div in potential:
                if div.find(string=re.compile(r'#')):
                    player_cards.append(div)
            print(f"🔍 Encontrados {len(player_cards)} cards por fallback")

        roster_data = []
        for card in player_cards:
            athlete = {
                'school_name': school_name,  # <-- garante que o nome da faculdade seja preenchido
                'jersey_number': '',
                'name': '',
                'position': '',
                'class_year': '',
                'height': '',
                'hometown': '',
                'high_school_club': ''
            }

            # --- Número ---
            number_elem = card.find(string=re.compile(r'^#'))
            if number_elem:
                num_text = number_elem.get_text(strip=True) if hasattr(number_elem, 'get_text') else number_elem
                match = re.search(r'#(\d+)', num_text)
                if match:
                    athlete['jersey_number'] = match.group(1)

            # --- Nome ---
            name_elem = card.find(['h3', 'h4'], class_=re.compile(r'name|title', re.I))
            if not name_elem:
                name_elem = card.find(class_=re.compile(r'player-name|full-name', re.I))
            if name_elem:
                raw_name = name_elem.get_text(strip=True)
                athlete['name'] = re.sub(r'Season\s+\d{4}-\d{2}', '', raw_name).strip()

            # --- Posição ---
            pos_elem = card.find(class_=re.compile(r'position', re.I))
            if pos_elem:
                athlete['position'] = pos_elem.get_text(strip=True)

            # ----- EXTRAÇÃO DE DETALHES (PRIORITÁRIA) -----
            # 1. Busca por elementos com classes específicas
            hometown_elem = card.find(class_=re.compile(r'hometown|city', re.I))
            if hometown_elem:
                athlete['hometown'] = hometown_elem.get_text(strip=True)
            highschool_elem = card.find(class_=re.compile(r'highschool|high-school|prev-school|previous-school', re.I))
            if highschool_elem:
                athlete['high_school_club'] = highschool_elem.get_text(strip=True)

            # 2. Bloco de detalhes estruturado (ul.roster__player-details)
            details_container = card.find(class_=re.compile(r'details|info|bio', re.I))
            if details_container:
                detail_items = details_container.find_all(['li', 'div', 'p'])
                for item in detail_items:
                    text = item.get_text(strip=True)
                    lower = text.lower()
                    if 'hometown:' in lower:
                        athlete['hometown'] = text.split(':', 1)[-1].strip()
                    elif 'high school:' in lower:
                        athlete['high_school_club'] = text.split(':', 1)[-1].strip()
                    elif 'previous school:' in lower:
                        athlete['high_school_club'] = text.split(':', 1)[-1].strip()
                    elif 'height:' in lower:
                        athlete['height'] = text.split(':', 1)[-1].strip()
                    elif 'class:' in lower or 'year:' in lower:
                        athlete['class_year'] = text.split(':', 1)[-1].strip()
                    elif '/' in text and not athlete['hometown'] and not athlete['high_school_club']:
                        left, right = split_hometown_school(text)
                        if left != text:
                            athlete['hometown'] = left
                            athlete['high_school_club'] = right
                    elif not athlete['height'] and re.search(r'\d+[\′\'"]\s*\d*[\′\'"]?', text):
                        athlete['height'] = re.search(r'\d+[\′\'"]\s*\d*[\′\'"]?', text).group(0)
                    elif not athlete['class_year'] and re.search(r'\b(Redshirt\s+)?(Fr\.?|So\.?|Jr\.?|Sr\.?|Gr\.?|Freshman|Sophomore|Junior|Senior|Graduate|Redshirt|Redshirt Freshman|Redshirt Sophomore|Redshirt Junior|Redshirt Senior|Graduate Student)\b', text, re.I):
                        athlete['class_year'] = re.search(r'\b(Redshirt\s+)?(Fr\.?|So\.?|Jr\.?|Sr\.?|Gr\.?|Freshman|Sophomore|Junior|Senior|Graduate|Redshirt|Redshirt Freshman|Redshirt Sophomore|Redshirt Junior|Redshirt Senior|Graduate Student)\b', text, re.I).group(0)

            # 3. Fallback: texto completo do card
            if (not athlete['hometown'] or not athlete['high_school_club'] or not athlete['height'] or not athlete['class_year']):
                card_text = card.get_text(separator='\n', strip=True)
                lines = [line.strip() for line in card_text.split('\n') if line.strip()]
                lines = [line for line in lines if line != athlete['name'] and line != athlete['position']]

                for line in lines:
                    lower = line.lower()
                    # Rótulos explícitos
                    if 'hometown:' in lower:
                        athlete['hometown'] = line.split(':', 1)[-1].strip()
                    elif 'high school:' in lower:
                        athlete['high_school_club'] = line.split(':', 1)[-1].strip()
                    elif 'previous school:' in lower:
                        athlete['high_school_club'] = line.split(':', 1)[-1].strip()
                    elif 'height:' in lower:
                        athlete['height'] = line.split(':', 1)[-1].strip()
                    elif 'class:' in lower or 'year:' in lower:
                        athlete['class_year'] = line.split(':', 1)[-1].strip()
                    # Separação por barra
                    elif '/' in line and not athlete['hometown'] and not athlete['high_school_club']:
                        left, right = split_hometown_school(line)
                        if left != line:
                            athlete['hometown'] = left
                            athlete['high_school_club'] = right
                    # Linhas compostas
                    elif any(k in line for k in ['lbs', 'Hometown', 'High School', 'Previous School']):
                        compound = extract_compound_line(line)
                        if compound['height']:
                            athlete['height'] = compound['height']
                        if compound['class_year']:
                            athlete['class_year'] = compound['class_year']
                        if compound['hometown']:
                            athlete['hometown'] = compound['hometown']
                        if compound['high_school_club']:
                            athlete['high_school_club'] = compound['high_school_club']
                    # Fallback regex
                    else:
                        if not athlete['height'] and re.search(r'\d+[\′\'"]\s*\d*[\′\'"]?', line):
                            athlete['height'] = re.search(r'\d+[\′\'"]\s*\d*[\′\'"]?', line).group(0)
                        if not athlete['class_year'] and re.search(r'\b(Redshirt\s+)?(Fr\.?|So\.?|Jr\.?|Sr\.?|Gr\.?|Freshman|Sophomore|Junior|Senior|Graduate|Redshirt|Redshirt Freshman|Redshirt Sophomore|Redshirt Junior|Redshirt Senior|Graduate Student)\b', line, re.I):
                            athlete['class_year'] = re.search(r'\b(Redshirt\s+)?(Fr\.?|So\.?|Jr\.?|Sr\.?|Gr\.?|Freshman|Sophomore|Junior|Senior|Graduate|Redshirt|Redshirt Freshman|Redshirt Sophomore|Redshirt Junior|Redshirt Senior|Graduate Student)\b', line, re.I).group(0)
                        if not athlete['hometown'] and re.search(r'[A-Z][a-zA-Z\s\.\-]+,\s*[A-Z]{2}', line):
                            athlete['hometown'] = re.search(r'[A-Z][a-zA-Z\s\.\-]+,\s*[A-Z]{2}', line).group(0)
                        if not athlete['hometown'] and re.search(r'[A-Z][a-zA-Z\s\.\-]+,\s*[A-Z][a-z]+', line):
                            athlete['hometown'] = re.search(r'[A-Z][a-zA-Z\s\.\-]+,\s*[A-Z][a-z]+', line).group(0)
                        if not athlete['high_school_club'] and any(k in lower for k in ['high school', 'academy', 'club', 'fc', 'sc', 'united', 'breakers', 'solar']):
                            athlete['high_school_club'] = line

            # Limpeza final: remove rótulos residuais
            for field in ['hometown', 'high_school_club', 'class_year', 'height']:
                if athlete[field]:
                    # Remove rótulos como "Hometown:", "High School:", etc.
                    athlete[field] = re.sub(r'^(hometown|high school|previous school|class|year|height)\s*:\s*', '', athlete[field], flags=re.I).strip()
                    # Remove palavra "Hometown" sozinha
                    if field == 'hometown' and athlete[field].lower().startswith('hometown'):
                        athlete[field] = athlete[field][8:].strip()
                    if field == 'high_school_club' and athlete[field].lower().startswith('high school'):
                        athlete[field] = athlete[field][11:].strip()
                    if field == 'high_school_club' and athlete[field].lower().startswith('previous school'):
                        athlete[field] = athlete[field][15:].strip()

            if athlete['name']:
                roster_data.append(athlete)

        print(f"✅ Total extraído: {len(roster_data)} atletas")
        return pd.DataFrame(roster_data)

    except Exception as e:
        print(f"❌ Erro no requests: {e}")
        return pd.DataFrame()

# ======================== ORQUESTRADOR PRINCIPAL ========================
def scrape_wmt_roster(url, school_name):
    if 'gostanford.com' in url:
        return scrape_stanford(url, school_name)
    else:
        return scrape_generic_wmt(url, school_name)

# ======================== EXECUÇÃO ========================
if __name__ == "__main__":
    try:
        import google.colab
        print("📦 Ambiente Colab detectado. Instalando chromedriver...")
        !apt-get update
        !apt install chromium-chromedriver
        !cp /usr/lib/chromium-browser/chromedriver /usr/bin
        !pip install selenium
    except ImportError:
        pass

    print("📂 Carregando arquivo de plataformas...")
    if not os.path.exists(CSV_FILE):
        print(f"❌ Arquivo não encontrado: {CSV_FILE}")
        exit(1)

    df = pd.read_csv(CSV_FILE)
    if 'platform' in df.columns:
        df_wmt = df[df['platform'] == 'WMT Digital']
    else:
        print("⚠️ Coluna 'platform' não encontrada. Processando todas as linhas.")
        df_wmt = df

    print(f"Escolas a processar: {len(df_wmt)}")
    failures = []

    for idx, row in df_wmt.iterrows():
        school = row['school_name']  # verifique se a coluna se chama 'school_name'
        gender = row.get('gender', '')
        roster_url = row.get('roster_url', '')

        if pd.isna(roster_url) or not str(roster_url).startswith('http'):
            print(f"\n[{idx+1}] {school} ({gender}) - Sem URL válida")
            failures.append({'school': school, 'gender': gender, 'reason': 'No roster URL', 'url': roster_url})
            continue

        print(f"\n[{idx+1}] {school} ({gender})")
        print(f"   URL: {roster_url}")

        df_result = scrape_wmt_roster(roster_url, school)

        if df_result is not None and not df_result.empty:
            df_result['gender'] = gender
            append_to_roster_csv(df_result, SUCCESS_FILE)
            print(f"   ✅ Sucesso: {len(df_result)} atletas")
        else:
            print(f"   ⚠️ Nenhum dado retornado")
            failures.append({'school': school, 'gender': gender, 'reason': 'No data found', 'url': roster_url})

        time.sleep(2)

    if failures:
        fail_df = pd.DataFrame(failures)
        if os.path.exists(FAILURE_FILE):
            fail_df.to_csv(FAILURE_FILE, mode='a', header=False, index=False, encoding='utf-8')
        else:
            fail_df.to_csv(FAILURE_FILE, index=False, encoding='utf-8')
        print(f"💾 Falhas registradas em {FAILURE_FILE}")
```

```python
# -*- coding: utf-8 -*-
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
import os
import time

# ======================== CONFIGURAÇÕES ========================
BASE_PATH = '/content/drive/MyDrive/RA_Project_EconDistance/raw_data/'
CSV_FILE = os.path.join(BASE_PATH, 'school_website_platforms.csv')
SUCCESS_FILE = os.path.join(BASE_PATH, 'soccer_rosters.csv')
FAILURE_FILE = os.path.join(BASE_PATH, 'soccer_roster_failures.csv')

# ======================== FUNÇÕES AUXILIARES ========================
def append_to_roster_csv(new_df, filepath, key_columns=['school_name', 'name']):
    """Adiciona novos dados ao CSV evitando duplicatas."""
    os.makedirs(os.path.dirname(filepath), exist_ok=True)
    # Reordena colunas para consistência
    column_order = ['school_name', 'gender', 'jersey_number', 'name', 'position', 'class_year', 'height', 'hometown', 'high_school_club']
    new_df = new_df[[col for col in column_order if col in new_df.columns]]
    if os.path.exists(filepath):
        existing_df = pd.read_csv(filepath)
        combined = pd.concat([existing_df, new_df], ignore_index=True)
        combined = combined.drop_duplicates(subset=key_columns, keep='last')
    else:
        combined = new_df
    combined.to_csv(filepath, index=False, encoding='utf-8')
    print(f"💾 Arquivo salvo em: {filepath}")

def scrape_prestosports_roster(url, school_name, gender):
    """
    Extrai dados do roster de sites PrestoSports.
    A página contém os dados em uma estrutura de lista.
    """
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36',
        'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9',
    }
    roster_data = []

    try:
        print(f"🌐 Acessando: {url}")
        response = requests.get(url, headers=headers, timeout=30)
        response.raise_for_status()
        soup = BeautifulSoup(response.content, 'html.parser')

        # Encontrar a lista de jogadores
        players_container = soup.find('ul', class_='sidearm-roster-players')
        if not players_container:
            print("❌ Não foi possível encontrar o container de jogadores.")
            return pd.DataFrame()

        players = players_container.find_all('li', class_='sidearm-roster-player')
        print(f"✅ Encontrados {len(players)} jogadores")

        for player in players:
            athlete = {
                'school_name': school_name,
                'gender': gender,
                'jersey_number': '',
                'name': '',
                'position': '',
                'class_year': '',
                'height': '',
                'hometown': '',
                'high_school_club': ''
            }

            # Número da camisa
            number_elem = player.find('span', class_='sidearm-roster-player-jersey-number')
            if number_elem:
                athlete['jersey_number'] = number_elem.get_text(strip=True)

            # Nome (dentro de h3 > a)
            name_elem = player.find('h3')
            if name_elem:
                name_link = name_elem.find('a')
                if name_link:
                    athlete['name'] = name_link.get_text(strip=True)
                else:
                    athlete['name'] = name_elem.get_text(strip=True)

            # Posição (dentro de .sidearm-roster-player-position .text-bold)
            pos_elem = player.find('span', class_='text-bold')
            if pos_elem:
                athlete['position'] = pos_elem.get_text(strip=True)

            # Altura
            height_elem = player.find('span', class_='sidearm-roster-player-height')
            if height_elem:
                athlete['height'] = height_elem.get_text(strip=True)

            # Ano acadêmico (class year)
            year_elem = player.find('span', class_='sidearm-roster-player-academic-year')
            if year_elem:
                athlete['class_year'] = year_elem.get_text(strip=True)
            else:
                class_div = player.find('div', class_='sidearm-roster-player-class-hometown')
                if class_div:
                    year_span = class_div.find('span', class_='sidearm-roster-player-academic-year')
                    if year_span:
                        athlete['class_year'] = year_span.get_text(strip=True)

            # Hometown e High School/Club
            class_div = player.find('div', class_='sidearm-roster-player-class-hometown')
            if class_div:
                hometown_elem = class_div.find('span', class_='sidearm-roster-player-hometown')
                hs_elem = class_div.find('span', class_='sidearm-roster-player-highschool')
                if hometown_elem:
                    athlete['hometown'] = hometown_elem.get_text(strip=True)
                if hs_elem:
                    athlete['high_school_club'] = hs_elem.get_text(strip=True)

                # Se não separados, tentar o texto completo e dividir por "/"
                if not athlete['hometown'] and not athlete['high_school_club']:
                    full_text = class_div.get_text(separator=' ', strip=True)
                    if '/' in full_text:
                        parts = full_text.split('/', 1)
                        athlete['hometown'] = parts[0].strip()
                        athlete['high_school_club'] = parts[1].strip() if len(parts) > 1 else ''
                    else:
                        athlete['hometown'] = full_text
            else:
                hometown_elem = player.find('span', class_='sidearm-roster-player-hometown')
                if hometown_elem:
                    athlete['hometown'] = hometown_elem.get_text(strip=True)
                hs_elem = player.find('span', class_='sidearm-roster-player-highschool')
                if hs_elem:
                    athlete['high_school_club'] = hs_elem.get_text(strip=True)

            # Limpeza
            for key in athlete:
                if athlete[key]:
                    athlete[key] = re.sub(r'\s+', ' ', athlete[key]).strip()

            if athlete['name']:
                roster_data.append(athlete)
                print(f"  ✓ {athlete['name']} (#{athlete['jersey_number']}) - {athlete['position']} - {athlete['class_year']}")

        print(f"\n✅ Total extraído: {len(roster_data)} atletas")

    except Exception as e:
        print(f"❌ Erro: {e}")
        import traceback
        traceback.print_exc()

    return pd.DataFrame(roster_data)

# ======================== EXECUÇÃO PRINCIPAL ========================
if __name__ == "__main__":
    # Montar Google Drive (necessário no Colab)
    try:
        from google.colab import drive
        drive.mount('/content/drive')
        print("✅ Google Drive montado.")
    except ImportError:
        print("⚠️ Não está no Colab, assumindo que o Drive já está montado.")

    print("📂 Carregando arquivo de plataformas...")
    if not os.path.exists(CSV_FILE):
        print(f"❌ Arquivo não encontrado: {CSV_FILE}")
        exit(1)

    df = pd.read_csv(CSV_FILE)

    # Verifica se a coluna 'platform' existe
    if 'platform' not in df.columns:
        print("❌ Coluna 'platform' não encontrada no CSV.")
        exit(1)

    # Filtra apenas as linhas com plataforma 'PrestoSports'
    df_presto = df[df['platform'].str.lower() == 'prestosports'].copy()
    print(f"✅ Encontradas {len(df_presto)} linhas com plataforma PrestoSports.")

    if df_presto.empty:
        print("Nenhuma linha para processar. Encerrando.")
        exit(0)

    # Verifica se as colunas necessárias existem
    required_cols = ['school_name', 'gender', 'roster_url']
    for col in required_cols:
        if col not in df_presto.columns:
            print(f"❌ Coluna '{col}' não encontrada no CSV. Verifique o formato.")
            exit(1)

    failures = []
    all_success = []

    for idx, row in df_presto.iterrows():
        school = row['school_name']
        gender = row['gender']
        roster_url = row['roster_url']

        if pd.isna(roster_url) or not str(roster_url).startswith('http'):
            print(f"\n[{idx+1}] {school} ({gender}) - URL inválida ou ausente: {roster_url}")
            failures.append({'school_name': school, 'gender': gender, 'url': roster_url, 'reason': 'Invalid URL'})
            continue

        print(f"\n[{idx+1}] Processando {school} ({gender})")
        print(f"   URL: {roster_url}")

        df_result = scrape_prestosports_roster(roster_url, school, gender)

        if df_result is not None and not df_result.empty:
            all_success.append(df_result)
            print(f"   ✅ Sucesso: {len(df_result)} atletas")
        else:
            print(f"   ⚠️ Nenhum dado retornado")
            failures.append({'school_name': school, 'gender': gender, 'url': roster_url, 'reason': 'No data extracted'})

        time.sleep(2)  # Pausa entre requisições

    # Salvar todos os resultados combinados
    if all_success:
        final_df = pd.concat(all_success, ignore_index=True)
        append_to_roster_csv(final_df, SUCCESS_FILE)
        print(f"\n✅ Total de atletas salvos: {len(final_df)}")
    else:
        print("\n⚠️ Nenhum dado foi extraído com sucesso.")

    # Salvar falhas
    if failures:
        fail_df = pd.DataFrame(failures)
        if os.path.exists(FAILURE_FILE):
            fail_df.to_csv(FAILURE_FILE, mode='a', header=False, index=False, encoding='utf-8')
        else:
            fail_df.to_csv(FAILURE_FILE, index=False, encoding='utf-8')
        print(f"💾 Falhas registradas em {FAILURE_FILE}")

    print("\n🏁 Processamento concluído.")
```
